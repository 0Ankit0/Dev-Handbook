### 4.1 The Primary Documents: Annual Reports, Quarterly Reports, and NEPSE Filings

Financial statements are the primary evidence base for any serious NEPSE decision because they reveal how the business earns, spends, finances, and reinvests cash. Pay particular attention to note disclosures, contingent liabilities, provisioning, capital plans, and management explanations that can materially change the headline numbers.

**NEPSE example:** Apply this idea to Nepal Telecom annual and quarterly reports and ask what the market is pricing correctly, what may still be misunderstood, and what evidence would change your view.

- Focus on the local disclosure, liquidity, and sector context before drawing a conclusion.
- Prefer a simple repeatable checklist over a prediction-heavy story.
- Revisit the idea after new NEPSE notices, quarterly numbers, or turnover shifts appear.
