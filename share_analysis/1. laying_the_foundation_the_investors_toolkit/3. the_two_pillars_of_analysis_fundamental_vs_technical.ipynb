### 3.1 What is Fundamental Analysis? The "What to Buy"

Fundamental analysis asks whether a business deserves your capital based on earnings power, balance-sheet strength, cash generation, governance, and future prospects. In a NEPSE setting, that means reading annual reports carefully and testing whether the story is supported by numbers rather than rumors.
