## 2.1 Variables and Memory

A **variable** in Python is not a box that holds data; rather, it is a **name** (or label) that points to an object in memory. This distinction is critical because it explains how Python handles assignment and modification.

### Variable Assignment

In Python, you create a variable by assigning a value to a name using the equals sign `=`.

```python
# Creating variables
user_name: str = "Alice"
age: int = 28
pi_value: float = 3.14159
is_active: bool = True
```

**Key Concepts:**
*   **Declaration and Assignment**: Python is dynamically typed. You do not need to declare the type beforehand (though we add type hints `: str`, `: int` for clarity and toolingâ€”more on this later).
*   **Naming Conventions (PEP 8)**: 
    *   Use `snake_case` for variable names (lowercase with underscores).
    *   Names must start with a letter or underscore (`_`), never a number.
    *   Names are case-sensitive (`User` and `user` are different variables).
    *   Avoid single letters like `x` or `i` except in mathematical contexts or short loops.

### How Python Manages Memory

When you write `user_name = "Alice"`, Python performs three steps:
1.  Creates a **string object** containing the value `"Alice"` in memory.
2.  Creates the variable name `user_name` in the namespace.
3.  **Binds** the name `user_name` to the object `"Alice"`.

Let's visualize this with a more complex example:

```python
# Assignment creates a reference, not a copy
original_list: list[int] = [1, 2, 3]
new_reference = original_list  # Both point to the same object

new_reference.append(4)

print(original_list)  # Output: [1, 2, 3, 4]
print(new_reference)  # Output: [1, 2, 3, 4]
```

**Explanation**: Both variables point to the same list object in memory. Modifying one modifies the other because they share the reference. This is a common source of bugs for beginners.

To create an actual copy, you would use:
```python
actual_copy = original_list.copy()  # or list(original_list) or original_list[:]
```

### Variable Reassignment
Variables are dynamic. You can rebind a name to a different type or value at any time:

```python
data = "Hello"      # data points to a string object
data = 42           # data now points to an integer object
data = [1, 2, 3]    # data now points to a list object
```

While Python allows this, **industry best practice** is to keep variable types consistent within a scope to avoid confusion and bugs.

## 2.2 Primitive Data Types

Python has several built-in data types. The "primitive" or basic types form the foundation of all data manipulation.

### Integers (`int`)
Integers are whole numbers, positive or negative, without decimals. Python integers have **unlimited precision** (they can be arbitrarily large, limited only by available memory).

```python
# Integers
positive: int = 42
negative: int = -17
zero: int = 0
large_number: int = 10**100  # 1 googol - Python handles this effortlessly

# Underscores for readability (PEP 515)
salary: int = 1_000_000  # Same as 1000000, easier to read
```

### Floating-Point Numbers (`float`)
Floats represent real numbers with decimal points. They follow the IEEE 754 double-precision standard (64-bit), which means they have precision limits.

```python
# Floats
temperature: float = 98.6
negative_float: float = -273.15
scientific: float = 3.5e5  # 3.5 Ã— 10^5 = 350000.0

# Precision Warning
result: float = 0.1 + 0.2
print(result)  # Output: 0.30000000000000004 (not exactly 0.3!)
```

**Industry Note**: Never use floats for precise financial calculations. Use the `decimal` module for money or use integers representing cents.

```python
from decimal import Decimal, getcontext
getcontext().prec = 6  # Set precision

price: Decimal = Decimal('0.10')
tax: Decimal = Decimal('0.20')
total: Decimal = price + tax
print(total)  # Output: 0.30 (exactly)
```

### Strings (`str`)
Strings are sequences of Unicode characters (text). In Python, strings are **immutable**, meaning once created, they cannot be changed.

```python
# String definitions
single_quotes: str = 'Hello'
double_quotes: str = "World"  # Preferred when text contains apostrophes
multiline: str = """This is a
multiline string that preserves
formatting and line breaks."""

# Unicode support
emoji: str = "Python ðŸ"
chinese: str = "ä½ å¥½ä¸–ç•Œ"

# Immutability demonstration
name: str = "Alice"
# name[0] = "B"  # This raises TypeError: 'str' object does not support item assignment
```

### Booleans (`bool`)
Booleans represent truth values: `True` or `False`. They are a subtype of integers, where `True` equals `1` and `False` equals `0`.

```python
is_logged_in: bool = True
has_permission: bool = False

# Boolean arithmetic (advanced but valid)
print(True + True)  # Output: 2
```

### None Type (`NoneType`)
`None` represents the absence of a value or a null value. It is not zero, not an empty string, and not Falseâ€”it is nothingness.

```python
user_input: str | None = None  # Indicates variable exists but has no value yet

def function_with_no_return() -> None:
    pass

result = function_with_no_return()
print(result)  # Output: None
```

### Checking Types
Use the `type()` function to inspect an object's type, and `isinstance()` for type checking (preferred in professional code):

```python
value: int = 42

print(type(value))              # Output: <class 'int'>
print(isinstance(value, int))   # Output: True
print(isinstance(value, str))   # Output: False

# Checking multiple types
print(isinstance(value, (int, float)))  # Output: True (is it int OR float?)
```

## 2.3 Operators

Operators are special symbols that perform operations on variables and values. Python categorizes them based on their function.

### Arithmetic Operators
Standard mathematical operations.

| Operator | Description | Example |
|----------|-------------|---------|
| `+` | Addition | `5 + 3` â†’ `8` |
| `-` | Subtraction | `5 - 3` â†’ `2` |
| `*` | Multiplication | `5 * 3` â†’ `15` |
| `/` | Division (always returns float) | `5 / 2` â†’ `2.5` |
| `//` | Floor Division (integer division) | `5 // 2` â†’ `2` |
| `%` | Modulus (remainder) | `5 % 2` â†’ `1` |
| `**` | Exponentiation | `2 ** 3` â†’ `8` |

```python
# Division behavior
regular_division: float = 17 / 5      # 3.4
floor_division: int = 17 // 5         # 3
remainder: int = 17 % 5               # 2

# Practical use case: checking even/odd
number: int = 42
is_even: bool = (number % 2 == 0)  # True if divisible by 2
```

### Comparison Operators
These compare values and return a Boolean (`True` or `False`).

| Operator | Meaning |
|----------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal to |
| `<=` | Less than or equal to |

```python
x: int = 10
y: int = 20

print(x == y)  # False
print(x != y)  # True
print(x < y)   # True

# Chained comparisons (Pythonic)
# Instead of: x < y and y < z
# You can write:
z: int = 30
print(x < y < z)  # True (checks if x < y AND y < z)
```

**Critical Warning**: Use `==` for value comparison, not `is`. The `is` operator checks identity (memory address), not equality.

```python
list_a: list[int] = [1, 2, 3]
list_b: list[int] = [1, 2, 3]

print(list_a == list_b)  # True (same values)
print(list_a is list_b)  # False (different objects in memory)
```

### Logical Operators
Combine conditional statements.

*   `and`: Returns `True` if both operands are true
*   `or`: Returns `True` if at least one operand is true  
*   `not`: Inverts the boolean value

```python
has_id: bool = True
is_over_18: bool = True

can_enter_club: bool = has_id and is_over_18  # True
has_permission: bool = has_id or is_over_18   # True (short-circuits if first is True)
is_minor: bool = not is_over_18               # False

# Truthiness: Non-zero numbers and non-empty collections are "truthy"
name: str = "Alice"
if name:  # Equivalent to: if name != ""
    print("Name exists")
```

### Bitwise Operators
Work on bits (binary representation). Essential for systems programming, cryptography, and performance optimization.

| Operator | Name | Description |
|----------|------|-------------|
| `&` | AND | Sets bit to 1 if both bits are 1 |
| `\|` | OR | Sets bit to 1 if one of two bits is 1 |
| `^` | XOR | Sets bit to 1 if only one of two bits is 1 |
| `~` | NOT | Inverts all bits |
| `<<` | Left Shift | Shift left by pushing zeros in from the right |
| `>>` | Right Shift | Shift right by pushing copies of the leftmost bit in from the left |

```python
a: int = 5      # Binary: 101
b: int = 3      # Binary: 011

print(a & b)    # 1  (Binary: 001)
print(a | b)    # 7  (Binary: 111)
print(a ^ b)    # 6  (Binary: 110)
print(a << 1)   # 10 (Binary: 1010 - shifted left, multiplied by 2)
print(a >> 1)   # 2  (Binary: 10 - shifted right, divided by 2)
```

## 2.4 String Manipulation

Strings are ubiquitous in programming. Mastering string manipulation is essential for data cleaning, parsing, and user interface development.

### Indexing and Slicing
Strings are sequences, meaning they are ordered collections where each character has an index (position).

```python
text: str = "Python"

# Indexing (0-based)
first_char: str = text[0]   # 'P'
last_char: str = text[-1]   # 'n' (negative indexing counts from end)

# Slicing [start:stop:step]
# Start is inclusive, stop is exclusive
substring: str = text[1:4]      # 'yth' (indices 1, 2, 3)
first_three: str = text[:3]     # 'Pyt' (start to index 2)
last_three: str = text[-3:]     # 'hon' (3rd from end to finish)
every_second: str = text[::2]   # 'Pto' (every 2nd character)
reversed_text: str = text[::-1] # 'nohtyP' (negative step reverses)
```

### String Methods (Immutability in Action)
Since strings are immutable, methods return new strings rather than modifying the original.

```python
message: str = "  Hello, World!  "

# Cleaning
clean: str = message.strip()        # "Hello, World!" (removes whitespace)
left_clean: str = message.lstrip()  # "Hello, World!  "
right_clean: str = message.rstrip() # "  Hello, World!"

# Case manipulation
lower: str = message.lower()        # "  hello, world!  "
upper: str = message.upper()        # "  HELLO, WORLD!  "
title_case: str = "hello world".title()  # "Hello World"

# Searching
position: int = message.find("World")  # 9 (index where "World" starts)
count: int = message.count("l")        # 3 (number of 'l's)

# Replacement (returns new string)
new_message: str = message.replace("World", "Python")  # "  Hello, Python!  "
```

### String Concatenation and Formatting

**Avoid this (inefficient for many operations):**
```python
# Bad: Creates many temporary string objects in memory
result: str = ""
for i in range(1000):
    result += str(i)  # O(nÂ²) complexity due to string immutability
```

**Do this instead (using join):**
```python
# Good: Efficient O(n) operation
parts: list[str] = [str(i) for i in range(1000)]
result: str = "".join(parts)
```

### Modern String Formatting: F-Strings (The Industry Standard)
Introduced in Python 3.6, f-strings (formatted string literals) are the most readable and performant way to embed expressions inside string literals.

```python
name: str = "Alice"
age: int = 30
pi: float = 3.14159

# Basic usage
greeting: str = f"Hello, {name}!"

# Expressions inside braces
info: str = f"Next year, {name} will be {age + 1} years old."

# Formatting numbers
formatted_pi: str = f"Pi rounded: {pi:.2f}"  # 3.14
padding: str = f"Age: {age:05d}"             # 00030 (zero-padding)
percentage: str = f"Success rate: {0.85:.1%}" # 85.0%

# Alignment
left: str = f"|{name:<10}|"   # |Alice     | (left align)
right: str = f"|{name:>10}|"  # |     Alice| (right align)
center: str = f"|{name:^10}|" # |  Alice   | (center align)
```

**Legacy Methods (for reference only):**
*   `%` formatting (printf-style): `"Hello, %s" % name` (Deprecated style)
*   `.format()` method: `"Hello, {}".format(name)` (Still valid but f-strings preferred)

## 2.5 Type Conversion (Casting)

Often, you need to convert data from one type to another. Python provides built-in functions for explicit conversion.

### Explicit Conversion
You manually convert types using constructor functions:

```python
# String to Integer
user_input: str = "42"
number: int = int(user_input)  # 42 (integer)

# Integer to String
str_number: str = str(42)      # "42" (string)

# String to Float
price: float = float("19.99")  # 19.99

# Number to Boolean (0 is False, non-zero is True)
bool_zero: bool = bool(0)      # False
bool_number: bool = bool(42)   # True

# Empty collections to Boolean
bool_empty: bool = bool("")    # False
bool_full: bool = bool("hi")   # True
```

### Handling Conversion Errors
Converting invalid strings raises `ValueError`. Always validate input in production code:

```python
data: str = "not_a_number"

try:
    value: int = int(data)
except ValueError:
    print(f"Error: Cannot convert '{data}' to integer")
    value = 0  # Default value
```

### Implicit Conversion (Type Coercion)
Python sometimes automatically converts types during operations, but it is conservative to avoid data loss:

```python
# Integer + Float = Float (widening conversion)
result: float = 5 + 3.0  # 8.0 (int promoted to float)

# Boolean in arithmetic (True=1, False=0)
calculation: int = 10 + True  # 11

# String + Number = Error (Python refuses to guess)
# bad = "5" + 5  # TypeError: can only concatenate str (not "int") to str

# Must explicitly convert
good: str = "5" + str(5)  # "55"
sum_result: int = int("5") + 5  # 10
```
