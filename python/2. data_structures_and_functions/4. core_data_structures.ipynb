## 📝 Part 1: Lists — Your Ordered, Changeable Collection

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.

### The Analogy

Think of a **shopping list on a whiteboard**. You can:
- ✅ Add new items at any time (`append`)
- ✅ Erase or cross off items (`remove`, `pop`)
- ✅ Reorder the list (`sort`, `reverse`)
- ✅ Check what's at position 1, 2, 3... (indexing)

A Python **list** works exactly the same way.

### Syntax & First Usage

Create a list with **square brackets `[ ]`**, separating items with commas.

````python
# Creating a list — square brackets [ ]
guests = ["Alice", "Bob", "Carol", "David"]

print("Guest list:", guests)
print("Number of guests:", len(guests))  # len() counts items
print("Type:", type(guests))
````

### 🔢 Indexing — Accessing Individual Items

Python uses **zero-based indexing** — count positions starting from 0, like slots:

```
guests = ["Alice", "Bob", "Carol", "David"]
slot:        0       1       2        3
```

> 💡 **Why start at 0?** It's a computing convention tied to memory addressing. You'll find it natural very quickly.

You can also count **backwards** using negative indices:

```
guests = ["Alice", "Bob", "Carol", "David"]
slot:       -4      -3      -2       -1
```

````python
guests = ["Alice", "Bob", "Carol", "David"]

# Positive indexing — counting from the front
print("First guest  (index  0):", guests[0])
print("Second guest (index  1):", guests[1])
print("Last guest   (index  3):", guests[3])

# Negative indexing — counting from the back
print()
print("Last guest         (index -1):", guests[-1])   # same as guests[3]
print("Second to last     (index -2):", guests[-2])
````

### ✂️ Slicing — Grabbing a Sub-List

Slicing is like **cutting out a section** of your list.

```python
my_list[start : stop : step]
```

- `start` — index to begin at (inclusive). Default: 0
- `stop` — index to end at (**exclusive** — item at `stop` is NOT included). Default: end of list
- `step` — how many positions to jump. Default: 1

````python
guests = ["Alice", "Bob", "Carol", "David", "Eve"]
#  index:    0       1       2        3       4

# guests[1:3] → from index 1, up to (not including) index 3
print("guests[1:3]  →", guests[1:3])   # ['Bob', 'Carol']

print("guests[:3]   →", guests[:3])    # first 3  (start omitted → 0)
print("guests[2:]   →", guests[2:])    # from index 2 to end
print("guests[-2:]  →", guests[-2:])   # last 2
print("guests[::2]  →", guests[::2])   # every other item (step=2)
print("guests[::-1] →", guests[::-1])  # reversed!
````

### 🔧 Common List Methods

A **method** is a built-in action you call on an object using dot notation: `object.method()`

````python
shopping = ["milk", "eggs", "bread"]
print("Start:", shopping)

# append — add to the END
shopping.append("butter")
print("append('butter')  →", shopping)

# insert — add at a SPECIFIC position
shopping.insert(1, "coffee")   # insert at index 1
print("insert(1,'coffee') →", shopping)

# remove — delete by VALUE (removes first match)
shopping.remove("eggs")
print("remove('eggs')    →", shopping)

# pop — remove and RETURN the item at an index (default: last item)
last_item = shopping.pop()
print("pop()             →", last_item, "| list now:", shopping)

# pop with index
second = shopping.pop(1)
print("pop(1)            →", second, "| list now:", shopping)
````

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5]
print("Original:", numbers)

# sort — rearranges in place (ascending by default)
numbers.sort()
print("sort()          →", numbers)

# reverse — flips the list in place
numbers.reverse()
print("reverse()       →", numbers)

# count — how many times a value appears
print("count(5)        →", numbers.count(5))

# index — find the position of the first occurrence of a value
print("index(9)        →", numbers.index(9))

# Sorting WITHOUT modifying the original: sorted()
original = [3, 1, 4, 1, 5]
new_sorted = sorted(original)
print("\nsorted() →", new_sorted, " original:", original)  # original unchanged!
````

## 📌 Part 2: Tuples — Your Ordered, Fixed Collection

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

### The Analogy

Think of **GPS coordinates**: `(37.7749, -122.4194)` is San Francisco. Those coordinates don't change — if you edited them accidentally, you'd end up in the wrong city!

A **tuple** is a list that *cannot be modified after creation*. This property is called **immutability**.

### Syntax & First Usage

Create a tuple with **parentheses `( )`**:

````python
# Creating a tuple — parentheses ( )
coordinates = (37.7749, -122.4194)
schedule    = ("9:00 AM", "10:30 AM", "12:00 PM", "3:00 PM")
rgb_red     = (255, 0, 0)       # RGB color — shouldn't change!

# ⚠️ Single-item tuple needs a trailing comma
single = (42,)   # without the comma, Python sees it as just (42) — not a tuple!
print("Single tuple:", single, "| type:", type(single))
print("Just parens: ", (42),  "| type:", type((42)))  # int, not tuple!

# Accessing elements — identical to lists
print("\nLatitude:", coordinates[0])
print("First slot:", schedule[0])
````

### 📦 Tuple Unpacking

One of Python's most elegant features — assign multiple variables from a tuple in one line:

````python
# Basic unpacking — number of variables must match number of items
city, country, population = ("Tokyo", "Japan", 13_960_000)
print(f"{city} is in {country} with {population:,} people")

# Practical: function returning multiple values
def get_screen_size():
    return (1920, 1080)   # returning a tuple is natural here

width, height = get_screen_size()
print(f"Screen: {width} x {height}")

# Swapping two variables — tuple unpacking makes this elegant
a, b = 10, 20
print(f"Before: a={a}, b={b}")
a, b = b, a
print(f"After:  a={a}, b={b}")
````

## ⚔️ vs: List vs Tuple

Sections like this become clearer when you separate capabilities from tradeoffs. As you read, ask what each option is optimized for and what complexity you accept in exchange.

> **The golden rule:** Use a **list** when your data will change. Use a **tuple** when your data is fixed.

````python
# ✅ LIST — mutable (changeable) — great for a shopping list
shopping_list = ["milk", "eggs", "bread"]
shopping_list.append("butter")    # perfectly fine
print("Updated list:", shopping_list)

# ✅ TUPLE — immutable (fixed) — great for days of the week
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri")
print("\nWeekdays:", weekdays)
print("First weekday:", weekdays[0])   # reading is fine
````

````python
# ❌ Trying to CHANGE a tuple raises an error
weekdays = ("Mon", "Tue", "Wed", "Thu", "Fri")

try:
    weekdays[0] = "Sunday"
except TypeError as e:
    print(f"TypeError: {e}")
    print()
    print("Immutability is a FEATURE, not a bug:")
    print("  ✅ Prevents accidental modification")
    print("  ✅ Can be used as dictionary keys (lists cannot!)")
    print("  ✅ Slightly more memory-efficient and faster")
````

````python
import sys

my_list  = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

print(f"List  memory: {sys.getsizeof(my_list)} bytes")
print(f"Tuple memory: {sys.getsizeof(my_tuple)} bytes")
print(f"Tuple is {sys.getsizeof(my_list) - sys.getsizeof(my_tuple)} bytes smaller!")
````

## 📖 Part 3: Dictionaries — Your Key-Value Lookup Table

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

### The Analogy

A **dictionary** works exactly like a real dictionary:
- You look up a **key** (the word) → you get a **value** (the definition)
- Or like a phone book: **name** → **phone number**

Every entry is a **key: value pair**. Keys must be unique.

### Syntax & First Usage

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

Create a dictionary with **curly braces `{ }`** and `key: value` pairs:

````python
# Creating a dictionary
phone_book = {
    "Alice": "555-1234",
    "Bob":   "555-5678",
    "Carol": "555-9012"
}

print("Phone book:", phone_book)
print("Type:", type(phone_book))

# Access by key — square brackets
print("\nAlice's number:", phone_book["Alice"])

# Add a new key-value pair
phone_book["David"] = "555-3456"
print("After adding David:", phone_book)

# Update an existing entry (same syntax as adding)
phone_book["Alice"] = "555-9999"
print("Alice's updated number:", phone_book["Alice"])

# Delete an entry
del phone_book["Bob"]
print("After deleting Bob:", phone_book)
````

### 🛡️ Safe Access with `.get()`

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

Accessing a missing key crashes with a `KeyError`. The `.get()` method is your safety net.

````python
phone_book = {"Alice": "555-1234", "Bob": "555-5678"}

# ❌ Direct access to missing key → crashes
try:
    print(phone_book["Eve"])
except KeyError as e:
    print(f"KeyError: {e}")

# ✅ .get() returns None instead of crashing
result = phone_book.get("Eve")
print(f"\n.get('Eve')           → {result}")

# ✅ .get() with a default value
result = phone_book.get("Eve", "Not found")
print(f".get('Eve','Not found') → {result}")

# ✅ 'in' operator to check before accessing
if "Eve" in phone_book:
    print(phone_book["Eve"])
else:
    print("\nEve is not in the phone book")
````

### 🔄 Iterating Over Dictionaries

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.



````python
meal_prefs = {
    "Alice": "vegetarian",
    "Bob":   "vegan",
    "Carol": "no restrictions",
    "David": "gluten-free"
}

# .keys() — iterate over keys
print("Guests:", list(meal_prefs.keys()))

# .values() — iterate over values
print("Preferences:", list(meal_prefs.values()))

# .items() — iterate over (key, value) pairs — most useful!
print("\nFull list:")
for name, pref in meal_prefs.items():
    print(f"  {name:8} → {pref}")
````

### 🪆 Nested Dictionaries

Dictionaries can hold other dictionaries — great for structured records like database rows.

````python
guests = {
    "Alice": {"age": 28, "meal": "vegetarian", "rsvp": True},
    "Bob":   {"age": 34, "meal": "vegan",      "rsvp": False},
    "Carol": {"age": 25, "meal": "none",        "rsvp": True}
}

# Access nested values by chaining keys
print("Alice's meal:", guests["Alice"]["meal"])
print("Bob RSVP'd:", guests["Bob"]["rsvp"])

# Iterate and filter
print("\nConfirmed guests:")
for name, info in guests.items():
    if info["rsvp"]:
        print(f"  ✅ {name} (age {info['age']}) — {info['meal']}")
````

## 🎒 Part 4: Sets — Your Unique-Items Collection

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

### The Analogy

Imagine a **bag of unique tokens**. If you try to put two identical tokens in, only *one* stays. That's a **set** — it automatically removes duplicates. Items in the bag don't have a fixed order.

### Syntax & First Usage

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.



````python
# Creating a set — curly braces { } WITHOUT colons
allergies = {"nuts", "gluten", "dairy", "nuts"}   # "nuts" appears twice!

print("Allergies set:", allergies)   # "nuts" appears only once
print("Type:", type(allergies))
print("Count:", len(allergies))

# Add an item
allergies.add("shellfish")
print("\nAfter add:", allergies)

# Remove an item (raises KeyError if missing — use .discard() for safe removal)
allergies.discard("dairy")       # safe — no error if not found
print("After discard:", allergies)

# Membership testing — extremely fast (O(1))
print("\n'nuts' in allergies:", "nuts" in allergies)
print("'soy'  in allergies:", "soy" in allergies)
````

### 🔗 Set Operations — The Math of Relationships

The important beginner shift here is moving from “it works on my machine” to “it behaves predictably in a real environment.” Each step matters because it reduces surprises when the software runs elsewhere.



````python
attending  = {"Alice", "Bob", "Carol", "David"}
paid       = {"Bob", "Carol", "Eve"}

print(f"Attending: {attending}")
print(f"Paid:      {paid}")
print()

# Union (|) — everyone in EITHER set
print("Union (|)      — in either:     ", attending | paid)

# Intersection (&) — only in BOTH
print("Intersection (&)— in both:      ", attending & paid)

# Difference (-) — in first but NOT in second
print("Difference (-)  — attending not paid:", attending - paid)

# Symmetric difference (^) — in one but NOT both
print("Sym. diff (^)  — in exactly one:", attending ^ paid)
````

````python
# 🏆 Sets' killer use-case: removing duplicates from a list

responses = ["Alice", "Bob", "Alice", "Carol", "Bob", "David", "Alice"]
print("Raw RSVP responses:", responses)
print(f"Total responses:    {len(responses)}")
print(f"Duplicates:         {len(responses) - len(set(responses))}")

unique = sorted(list(set(responses)))   # convert back to sorted list
print(f"Unique guests:      {unique}")
````

## ⚔️ vs: Dictionary vs Set — The `{}` Confusion

Both use curly braces `{}` — here's how to tell them apart:

````python
# Dictionary — curly braces WITH key: value pairs
my_dict = {"name": "Alice", "age": 28}
print("Dict:", my_dict, "→ type:", type(my_dict).__name__)

# Set — curly braces WITHOUT key: value (just values)
my_set = {"Alice", "Bob", "Carol"}
print("Set: ", my_set,  "→ type:", type(my_set).__name__)

# ⚠️ GOTCHA: empty {} is a DICT, not a set!
empty_curly = {}
print("\n{}   type:", type(empty_curly).__name__, "← dict, not set!")

# ✅ To create an empty set, use set()
empty_set = set()
print("set() type:", type(empty_set).__name__, "← correct!")
````

## ⚔️ vs: List Search vs Dictionary Lookup — Speed Matters

The hard part here is usually not the syntax but the boundary between similar ideas. Keep comparing the job each option does best, the tradeoff it introduces, and the clues that tell you which one fits the situation in front of you.

When Python searches a **list**, it scans items one by one — called **O(n)** ("order n") performance.  
When Python looks up a **dictionary key**, it jumps straight to the answer using *hashing* — **O(1)** ("order 1", constant time).

> 💡 **O(n)** means: the time grows with the size of the list.  
> **O(1)** means: the time stays the same regardless of size.

````python
import time

SIZE = 1_000_000
large_list = list(range(SIZE))
large_dict = {i: f"value_{i}" for i in range(SIZE)}

# Search in LIST — worst case: item at the very end
start = time.perf_counter()
_ = large_list.index(999_999)
list_ms = (time.perf_counter() - start) * 1000

# Search in DICT — instant hash lookup
start = time.perf_counter()
_ = large_dict[999_999]
dict_ms = (time.perf_counter() - start) * 1000

print(f"List search: {list_ms:.4f} ms")
print(f"Dict lookup: {dict_ms:.6f} ms")
print(f"Dict was ~{list_ms/dict_ms:.0f}x faster!")
print()
print("💡 Rule: If you need fast lookups, use a dictionary.")
print("   If you need to search by value repeatedly, consider a dict.")
````

### What If #1: What if you use a list as a dictionary key?

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.



````python
# Dictionary keys must be HASHABLE — they cannot change.
# Lists are mutable, so Python refuses to hash them.

try:
    bad_dict = {[1, 2, 3]: "list as key"}
except TypeError as e:
    print(f"TypeError: {e}")
    print()
    print("Why? Python 'hashes' dictionary keys for fast lookup.")
    print("A list can change, making its hash unpredictable.")
    print()
    print("✅ Fix: use a TUPLE (immutable) as the key:")
    good_dict = {(1, 2, 3): "tuple as key works!"}
    print("  ", good_dict)
````

### What If #2: What if you access a dict key that doesn't exist?

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.



````python
grades = {"Alice": 95, "Bob": 87}

# 1. Direct access → KeyError (program crashes)
print("--- Direct access ---")
try:
    print(grades["Carol"])
except KeyError as e:
    print(f"KeyError: {e}")

# 2. .get() → returns None (safe)
print("\n--- .get() ---")
print(".get('Carol')       →", grades.get("Carol"))
print(".get('Carol', 0)    →", grades.get("Carol", 0))   # custom default

# 3. 'in' check → safe and explicit
print("\n--- 'in' check ---")
if "Carol" not in grades:
    print("Carol has no grade yet — assigning 0")
    grades["Carol"] = 0
print("Grades:", grades)
````

| Structure | Syntax | Ordered | Mutable | Unique | Best For |

| **Set** | `{a, b, c}` | ❌ | ✅ | ✅ | Unique items, set math |
