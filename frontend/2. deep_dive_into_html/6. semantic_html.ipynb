## Introduction

The easiest way to read this section is to keep one plain-language question in view: what is Introduction actually responsible for? Once that job is clear, the terminology stops feeling arbitrary and the details start to line up.

Semantic HTML refers to the use of HTML elements that convey meaning about the structure and content of a web page, rather than just describing how content should look. In the early days of the web, developers often relied heavily on generic containers like `<div>` and `<span>` to structure content, resulting in markup that was difficult for machines to interpret.

Modern HTML5 introduces semantic elements that clearly describe their meaning to both the browser and the developer. Using semantic HTML is not merely a best practice—it is a fundamental requirement for building accessible, maintainable, and search-engine-friendly websites.

When you write semantic HTML, you create a document that is understandable by:
- **Assistive technologies** (screen readers for visually impaired users)
- **Search engines** (Google, Bing, DuckDuckGo)
- **Developer tools** (code editors, linters, validators)
- **Future developers** (including your future self)

In this chapter, you will learn how to structure documents meaningfully, choose appropriate semantic elements, and understand the real-world impact of semantic markup on accessibility and SEO.

### The Problem with Non-Semantic Markup

Before HTML5, developers used generic containers with class names to indicate meaning:


````python
<!-- Non-semantic markup (pre-HTML5 approach) -->
<div class="header">
    <div class="nav">
        <div class="nav-item"><a href="/">Home</a></div>
        <div class="nav-item"><a href="/about">About</a></div>
    </div>
</div>
<div class="main-content">
    <div class="article">
        <div class="article-header">
            <div class="title">Article Title</div>
        </div>
        <div class="article-body">
            <p>Article content here...</p>
        </div>
    </div>
</div>
<div class="footer">
    <div class="copyright">© 2024</div>
</div>
````

**Problems with this approach:**
- Screen readers cannot identify navigation landmarks
- Search engines struggle to identify the main content
- The document has no structural outline
- Maintenance requires understanding arbitrary class names

**Accessibility Example:**

A screen reader user navigates a page using landmarks. With semantic HTML:

```
User hears: "Banner landmark, heading level 1, Site Name"
User hears: "Navigation landmark, list with 5 items"
User hears: "Main landmark, article, heading level 2, Article Title"
```

Without semantics, they hear:
```
User hears: "Div. Div. Link, Home. Link, About. Div. Div. 
             Heading level 2, Article Title" (if headings are used at all)
```

**SEO Example:**

Google's algorithms specifically look for `<article>`, `<nav>`, and proper heading hierarchies to:
- Identify the main content vs. sidebars/footers

- Build knowledge graphs

## 6.2 Document Structure Elements

HTML5 introduced structural elements that define major regions of a document. These create "landmarks" that assistive technologies use for navigation.

### The `<header>` Element

The `<header>` element represents introductory content, typically containing:
- Heading elements (`<h1>`–`<h6>`)
- Logo or site identity
- Search form
- Navigation (though `<nav>` is often nested inside or separate)

**Usage Rules:**
- Can be used multiple times per page (site header, article header, section header)
- Cannot be nested inside `<footer>`, `<address>`, or another `<header>`
- Should not be confused with `<head>` (which contains metadata)


````python
<!-- Site header -->
<header>
    <a href="/" class="logo">Company Name</a>
    <nav>
        <ul>
            <li><a href="/">Home</a></li>
            <li><a href="/services">Services</a></li>
            <li><a href="/contact">Contact</a></li>
        </ul>
    </nav>
</header>

<!-- Article header -->
<article>
    <header>
        <h2>Understanding Semantic HTML</h2>
        <p class="byline">By Jane Developer | January 15, 2024</p>
    </header>
    <p>Article content here...</p>
</article>
````

**Industry Standard Pattern:**


````python
<body>
    <header>
        <!-- Site-wide header content -->
        <nav><!-- Primary navigation --></nav>
    </header>
    
    <main>
        <article>
            <header>
                <!-- Article-specific header -->
                <h1>Article Title</h1>
                <time datetime="2024-01-15">January 15, 2024</time>
            </header>
            <!-- Article content -->
        </article>
    </main>
</body>
````

### The `<nav>` Element

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.

The `<nav>` element represents a block of navigation links. Not all links need to be in `<nav>`—only major navigation blocks.

**When to use `<nav>`:**
- Primary site navigation
- Table of contents
- Pagination controls
- Breadcrumb navigation

**When NOT to use `<nav>`:**
- Footer links (privacy policy, terms of service) unless they constitute a major navigation block
- Individual links within text
- Social media links in a sidebar (unless it's the main navigation)


````python
<!-- Primary navigation -->
<nav aria-label="Main Navigation">
    <ul>
        <li><a href="/" aria-current="page">Home</a></li>
        <li><a href="/about">About</a></li>
        <li><a href="/services">Services</a></li>
        <li><a href="/contact">Contact</a></li>
    </ul>
</nav>

<!-- Breadcrumb navigation -->
<nav aria-label="Breadcrumb">
    <ol>
        <li><a href="/">Home</a></li>
        <li><a href="/blog">Blog</a></li>
        <li aria-current="page">Current Article</li>
    </ol>
</nav>

<!-- Table of contents -->
<nav aria-label="Table of Contents">
    <ol>
        <li><a href="#intro">Introduction</a></li>
        <li><a href="#methods">Methods</a></li>
        <li><a href="#results">Results</a></li>
        <li><a href="#conclusion">Conclusion</a></li>
    </ol>
</nav>
````

**Accessibility Note:** Always include an `aria-label` when you have multiple `<nav>` elements to distinguish them for screen reader users.
### The `<main>` Element

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.

The `<main>` element represents the dominant content of the `<body>` of a document. There must be only one `<main>` per page (excluding iframes), and it cannot be nested inside `<article>`, `<aside>`, `<footer>`, `<header>`, or `<nav>`.

**Critical for Accessibility:**
- Screen reader users can jump directly to main content
- Skip links often target the main element
- Search engines weight content inside `<main>` more heavily


````python
<body>
    <header>
        <nav><!-- Navigation --></nav>
    </header>
    
    <main>
        <h1>Welcome to Our Site</h1>
        <p>This is the primary content of the page...</p>
        
        <article>
            <h2>Latest News</h2>
            <!-- Article content -->
        </article>
    </main>
    
    <aside>
        <!-- Sidebar content -->
    </aside>
    
    <footer>
        <!-- Footer content -->
    </footer>
</body>
````

**Skip Link Pattern:**


````python
<body>
    <a href="#main-content" class="skip-link">Skip to main content</a>
    
    <header><!-- Header content --></header>
    
    <main id="main-content">
        <!-- Main content -->
    </main>
</body>

<style>
.skip-link {
    position: absolute;
    left: -9999px;
    z-index: 999;
    padding: 1em;
    background-color: #000;
    color: #fff;
    text-decoration: none;
}

.skip-link:focus {
    left: 50%;
    transform: translateX(-50%);
    top: 0;
}
</style>
````

### The `<footer>` Element

The `<footer>` element represents a footer for its nearest sectioning content or sectioning root element. A footer typically contains:
- Author information
- Copyright data
- Related documents
- Contact information
- Site maps

Like `<header>`, multiple footers can exist per page (site footer, article footer).


````python
<!-- Site footer -->
<footer>
    <p>&copy; 2024 Company Name. All rights reserved.</p>
    <address>
        Contact us at <a href="mailto:info@example.com">info@example.com</a>
    </address>
</footer>

<!-- Article footer -->
<article>
    <h2>Article Title</h2>
    <p>Content...</p>
    
    <footer>
        <p>Posted by <a href="/author/jane">Jane Developer</a></p>
        <p>Filed under: <a href="/category/tech">Technology</a></p>
    </footer>
</article>
````

### The `<aside>` Element

The `<aside>` element represents a portion of a document whose content is only indirectly related to the document's main content. Common uses include:
- Sidebars
- Call-out boxes
- Advertising
- Groups of nav elements
- Pull quotes


````python
<main>
    <article>
        <h1>Article Title</h1>
        <p>Main article content...</p>
        
        <aside class="pull-quote">
            <q>This is an important quote from the article.</q>
            <cite>— Author Name</cite>
        </aside>
        
        <p>More article content...</p>
    </article>
</main>

<aside class="sidebar">
    <h2>Related Articles</h2>
    <ul>
        <li><a href="/article-1">Related Article 1</a></li>
        <li><a href="/article-2">Related Article 2</a></li>
    </ul>
</aside>
````

## 6.3 Content Sectioning Elements

A first read goes better when you connect every new idea to a concrete responsibility and a concrete use case. Once those anchors are in place, the terminology in this section becomes much easier to reuse accurately.

### The `<article>` Element

The `<article>` element represents a complete, or self-contained, composition in a document, page, application, or site. This could be:
- A magazine or newspaper article
- A blog entry
- A product card
- A user-submitted comment
- An interactive widget
- A forum post

**Key characteristic:** An article should make sense on its own when syndicated (e.g., in an RSS feed).


````python
<!-- Blog post as article -->
<article>
    <header>
        <h2>Understanding CSS Grid</h2>
        <p class="meta">Published on <time datetime="2024-01-15">January 15, 2024</time></p>
    </header>
    
    <p>CSS Grid Layout is a two-dimensional layout system...</p>
    
    <footer>
        <p>Tags: <a href="/tag/css">CSS</a>, <a href="/tag/layout">Layout</a></p>
    </footer>
</article>

<!-- Product card as article -->
<article class="product-card">
    <img src="product.jpg" alt="Product Name">
    <h3>Product Name</h3>
    <p class="price">$29.99</p>
    <p class="description">Product description...</p>
    <button>Add to Cart</button>
</article>

<!-- Comment as article (nested) -->
<article>
    <h2>Blog Post Title</h2>
    <p>Post content...</p>
    
    <section class="comments">
        <h3>Comments</h3>
        
        <article class="comment">
            <header>
                <h4>Comment by John Doe</h4>
                <time datetime="2024-01-16">Jan 16, 2024</time>
            </header>
            <p>Great article! Thanks for sharing.</p>
        </article>
        
        <article class="comment">
            <header>
                <h4>Comment by Jane Smith</h4>
                <time datetime="2024-01-17">Jan 17, 2024</header>
            <p>I learned a lot from this.</p>
        </article>
    </section>
</article>
````

### The `<section>` Element

The `<section>` element represents a standalone section of a document, which doesn't have a more specific semantic element to represent it. Sections should always have a heading (with rare exceptions).

**Use `<section>` when:**
- You need a standalone section with a heading
- The content forms a distinct thematic grouping
- It would appear in the document outline

**Do NOT use `<section>` as:**
- A generic container (use `<div>`)
- A styling wrapper (use `<div>`)
- Simply to group content for JavaScript targeting (use `<div>`)


````python
<!-- CORRECT: Section with heading -->
<section>
    <h2>Introduction</h2>
    <p>Introductory content...</p>
</section>

<section>
    <h2>Methodology</h2>
    <p>How we conducted the research...</p>
</section>

<!-- INCORRECT: Section without heading -->
<section class="wrapper">  <!-- Use div instead -->
    <p>Just some text...</p>
</section>

<!-- INCORRECT: Section purely for styling -->
<section class="red-box">  <!-- Use div instead -->
    <p>Styled content...</p>
</section>
````

**Article vs Section:**


````python
<!-- Article contains sections -->
<article>
    <h1>Complete Guide to HTML</h1>
    
    <section>
        <h2>Getting Started</h2>
        <p>Content...</p>
    </section>
    
    <section>
        <h2>Advanced Techniques</h2>
        <p>Content...</p>
    </section>
</article>

<!-- Section contains articles -->
<section class="latest-posts">
    <h2>Latest Blog Posts</h2>
    
    <article>
        <h3>Post One</h3>
        <p>Summary...</p>
    </article>
    
    <article>
        <h3>Post Two</h3>
        <p>Summary...</p>
    </article>
</section>
````

**Rule of thumb:**
- Use `<article>` for content that could stand alone (syndicatable)
- Use `<section>` for thematic grouping of content within a larger piece
- If in doubt whether it needs a heading, it's probably not a section

### The `<hgroup>` Element

The `<hgroup>` element represents a heading and related content. It groups a heading with any secondary content like subheadings, alternative titles, or taglines.

**Note:** This element has a complicated history in the spec, but is now valid in HTML5.


````python
<hgroup>
    <h1>The Lord of the Rings</h1>
    <p>The Fellowship of the Ring</p>
</hgroup>

<hgroup>
    <h1>Dr. Strangelove</h1>
    <p>Or: How I Learned to Stop Worrying and Love the Bomb</p>
</hgroup>
````

### Heading Hierarchy

If this topic is new, keep translating each new term into a simple question: what job does it do, when would you reach for it, and what would you confuse it with if you were moving too quickly? That habit makes the section easier to retain.

Proper heading structure is crucial for accessibility and SEO. Headings (`<h1>` through `<h6>`) should represent the outline of your document.

**Rules:**
- One `<h1>` per page (typically the page title)
- Don't skip levels (h2 → h4 is bad)
- Headings create an outline; treat them as such
- Visual styling should match semantic importance (or be adjusted with CSS)


````python
<!-- CORRECT: Logical outline -->
<h1>Main Page Title</h1>
    <h2>Section One</h2>
        <h3>Subsection A</h3>
        <h3>Subsection B</h3>
    <h2>Section Two</h2>
        <h3>Subsection C</h3>
            <h4>Detail 1</h4>
            <h4>Detail 2</h4>

<!-- INCORRECT: Skipping levels -->
<h1>Main Title</h1>
    <h3>Section One</h3>  <!-- Skipped h2! -->
        <h5>Subsection</h5>  <!-- Skipped h4! -->

<!-- INCORRECT: Multiple h1s -->
<h1>Site Name</h1>
<h1>Article Title</h1>  <!-- Should be h2 or h1 in separate article -->
````

**Document Outline Algorithm:**

```
┌─────────────────────────────────────────────────────────────────┐
│                  DOCUMENT OUTLINE STRUCTURE                      │
│                                                                 │
│  h1: Main Page Title                                            │
│  ├── h2: Introduction                                           │
│  ├── h2: Methods                                                │
│  │   ├── h3: Data Collection                                    │
│  │   └── h3: Analysis                                           │
│  ├── h2: Results                                                │
│  │   └── h3: Statistical Findings                               │
│  └── h2: Conclusion                                             │
│                                                                 │
│  Screen readers can navigate this outline.                      │
│  Search engines use this for featured snippets.                 │
│                                                                 │
└─────────────────────────────────────────────────────────────────┘
```

**Sectioning content resets the outline:**


````python
<body>
    <h1>Site Name</h1>
    
    <article>
        <h1>Article Title</h1>  <!-- Valid: article creates new outline -->
        <p>Content...</p>
    </article>
    
    <article>
        <h1>Another Article</h1>  <!-- Also valid -->
        <p>Content...</p>
    </article>
</body>
````

## 6.4 Text Semantics

Beyond structural elements, HTML provides semantic elements for inline text content. These add meaning to words and phrases within your content.

### Quotations

**`<blockquote>`:** For long quotations (block-level)


````python
<blockquote cite="https://developer.mozilla.org/en-US/docs/Web/HTML">
    <p>HTML (HyperText Markup Language) is the most basic building block of the Web. It defines the meaning and structure of web content.</p>
</blockquote>

<!-- With citation -->
<figure>
    <blockquote>
        <p>The power of the Web is in its universality. Access by everyone regardless of disability is an essential aspect.</p>
    </blockquote>
    <figcaption>— Tim Berners-Lee, <cite>W3C</cite></figcaption>
</figure>
````

**`<q>`:** For short, inline quotations


````python
<p>As Tim Berners-Lee said, <q>The Web does not just connect machines, it connects people</q>.</p>
````

**`<cite>`:** For the title of a creative work (book, paper, song, movie) or the name of the author when used within a `<figcaption>`.


````python
<p>My favorite book is <cite>The Pragmatic Programmer</cite> by Andrew Hunt.</p>

<!-- Note: cite outside figcaption refers to the work, not the person -->
<figure>
    <blockquote>
        <p>Code is read much more often than it is written.</p>
    </blockquote>
    <figcaption>— <cite>Guido van Rossum</cite></figcaption>
</figure>
````

### Code and Technical Elements

**`<code>`:** For fragments of computer code (inline)


````python
<p>Use the <code>console.log()</code> function to debug your JavaScript.</p>
````

**`<pre>`:** For preformatted text (preserves whitespace)


````python
<pre>
function greeting() {
    console.log("Hello, World!");
    return true;
}
</pre>
````

**Combining `<pre>` and `<code>`:**


````python
<pre><code>function calculate(x, y) {
    return x + y;
}</code></pre>
````

**`<kbd>`:** For user keyboard input


````python
<p>Press <kbd>Ctrl</kbd> + <kbd>S</kbd> to save the file.</p>
<p>To copy, press <kbd>Cmd</kbd> + <kbd>C</kbd> on Mac.</p>
````

**`<samp>`:** For sample output from a computer program


````python
<p>The program output:</p>
<samp>Error: File not found
Please check the path and try again.</samp>
````

**`<var>`:** For variables in mathematical expressions or programming


````python
<p>The area of a circle is calculated using <var>πr²</var>, 
where <var>r</var> is the radius.</p>
````

### Contact Information

**`<address>`:** For contact information (not for arbitrary addresses)


````python
<footer>
    <address>
        <p>Company Name</p>
        <p>123 Main Street<br>
        City, State 12345</p>
        <p>Email: <a href="mailto:contact@example.com">contact@example.com</a></p>
        <p>Tel: <a href="tel:+1234567890">(123) 456-7890</a></p>
    </address>
</footer>

<!-- For article author -->
<article>
    <footer>
        <address>
            Written by <a href="/author/jane">Jane Developer</a><br>
            Visit me at: <a href="https://example.com">Example.com</a>
        </address>
    </footer>
</article>
````

**Important:** Do not use `<address>` for arbitrary street addresses that aren't contact information for the content.

### Time and Dates

**`<time>`:** For representing dates and times in a machine-readable format


````python
<!-- Without datetime attribute (must be a valid date/time string) -->
<time>January 15, 2024</time>

<!-- With datetime attribute (recommended) -->
<time datetime="2024-01-15">January 15, 2024</time>
<time datetime="2024-01-15T14:30">January 15 at 2:30 PM</time>
<time datetime="14:30:00">2:30 PM</time>
<time datetime="2024-W03">Week 3 of 2024</time>

<!-- Duration -->
<time datetime="PT2H30M">2 hours 30 minutes</time>
````

**Benefits:**
- Search engines can extract event dates
- Calendar applications can add events
- Assistive technologies can pronounce dates correctly

### Figures and Captions

**`<figure>`:** Self-contained content (images, diagrams, code, quotes) referenced from the main content

**`<figcaption>`:** Caption or legend for the figure


````python
<figure>
    <img src="chart.png" alt="Bar chart showing sales growth">
    <figcaption>Sales growth increased by 150% in Q4 2023.</figcaption>
</figure>

<figure>
    <pre><code>&lt;article&gt;
    &lt;h1&gt;Title&lt;/h1&gt;
    &lt;p&gt;Content&lt;/p&gt;
&lt;/article&gt;</code></pre>
    <figcaption>Semantic HTML structure example.</figcaption>
</figure>

<figure>
    <blockquote>
        <p>Simplicity is the ultimate sophistication.</p>
    </blockquote>
    <figcaption><cite>Leonardo da Vinci</cite></figcaption>
</figure>
````

**Key points:**
- `<figcaption>` must be first or last child of `<figure>`
- Figure content is self-contained (if moved away, it still makes sense)
- Multiple images can share one caption

## 6.5 Lists Revisited

While covered in Chapter 2, lists deserve deeper exploration for semantic usage patterns.

### Navigation Menus

Navigation should use unordered lists because the order of items doesn't change the meaning:


````python
<nav aria-label="Main Navigation">
    <ul>
        <li><a href="/">Home</a></li>
        <li><a href="/about">About</a></li>
        <li><a href="/services">Services</a></li>
        <li><a href="/contact">Contact</a></li>
    </ul>
</nav>
````

**Why lists for navigation:**
- Screen readers announce "List with X items"
- Users know how many items to expect
- Provides structural meaning (these items are related)
- Easy to style consistently

### Breadcrumb Navigation

Breadcrumbs show the path to the current page. Use an ordered list because the sequence matters:


````python
<nav aria-label="Breadcrumb">
    <ol class="breadcrumb">
        <li><a href="/">Home</a></li>
        <li><a href="/products">Products</a></li>
        <li><a href="/products/electronics">Electronics</a></li>
        <li aria-current="page">Laptops</li>
    </ol>
</nav>
````

**Styling breadcrumbs:**


````python
.breadcrumb {
    list-style: none;
    display: flex;
    gap: 0.5rem;
}

.breadcrumb li:not(:last-child)::after {
    content: "›";
    margin-left: 0.5rem;
    color: #666;
}

.breadcrumb [aria-current="page"] {
    color: #000;
    text-decoration: none;
    font-weight: bold;
}
````

### Definition Lists

**Styling definition lists:**


````python
.product-details {
    display: grid;
    grid-template-columns: auto 1fr;
    gap: 0.5rem 1rem;
}

.product-details dt {
    font-weight: bold;
    color: #666;
}

.product-details dd {
    margin: 0;
}
````

## 6.6 Semantic vs Non-Semantic Elements

Sections like this become clearer when you separate capabilities from tradeoffs. As you read, ask what each option is optimized for and what complexity you accept in exchange.

### When to Use `<div>`

The `<div>` element is a generic container with no semantic meaning. Use it when:
- No other semantic element is appropriate
- You need a styling wrapper
- Grouping elements purely for CSS or JavaScript hooks

### When to Use `<span>`

The `<span>` element is an inline generic container. Use it when:
- You need to style a portion of text inline

### Avoiding "Divitis"

**Divitis** is the overuse of `<div>` elements when semantic alternatives exist.

**Before (Divitis):**


````python
<div class="header">
    <div class="logo">Site Name</div>
    <div class="nav">
        <div class="nav-item"><a href="/">Home</a></div>
    </div>
</div>
<div class="main">
    <div class="article">
        <div class="title">Article Title</div>
        <div class="content">Content...</div>
    </div>
</div>
<div class="footer">Copyright 2024</div>
````

**After (Semantic):**


````python
<header>
    <h1>Site Name</h1>
    <nav>
        <ul>
            <li><a href="/">Home</a></li>
        </ul>
    </nav>
</header>
<main>
    <article>
        <h2>Article Title</h2>
        <p>Content...</p>
    </article>
</main>
<footer><p>Copyright 2024</p></footer>
````

**Benefits of semantic approach:**
- Reduced HTML size (fewer classes needed)
- Better accessibility (landmarks, headings)
- Improved SEO
- Easier maintenance
- Better default styling (browsers have default styles for semantic elements)

### Semantic HTML Checklist

Before using a `<div>` or `<span>`, ask:

1. Is this a page header? → Use `<header>`
2. Is this navigation? → Use `<nav>`
3. Is this the main content? → Use `<main>`
4. Is this an article/blog post/comment? → Use `<article>`
5. Is this a section with a heading? → Use `<section>`
6. Is this a sidebar/tangent content? → Use `<aside>`
7. Is this a footer? → Use `<footer>`
8. Is this emphasized text? → Use `<em>` or `<strong>`
9. Is this code? → Use `<code>`
10. Is this a time/date? → Use `<time>`

Only after answering "no" to all should you use `<div>` or `<span>`.

## Chapter Summary

In this chapter, you learned the importance of semantic HTML for creating accessible, SEO-friendly, and maintainable websites:

1. **Document Structure**: Use `<header>`, `<nav>`, `<main>`, `<aside>`, and `<footer>` to create page landmarks that help users navigate.

2. **Content Sectioning**: Use `<article>` for self-contained content, `<section>` for thematic groupings with headings, and understand that these elements create document outlines.

3. **Text Semantics**: Use `<blockquote>` and `<q>` for quotations, `<code>`, `<pre>`, `<kbd>`, and `<samp>` for technical content, `<address>` for contact info, and `<time>` for dates.

4. **Lists**: Use unordered lists (`<ul>`) for navigation, ordered lists (`<ol>`) for sequences like breadcrumbs, and definition lists (`<dl>`) for key-value pairs.

5. **Avoiding Divitis**: Reserve `<div>` and `<span>` for when no semantic element exists, using them as styling hooks rather than structural elements.

### Key Takeaways

- Semantic HTML is the foundation of web accessibility
- Screen readers rely on landmarks (`<main>`, `<nav>`, etc.) and headings for navigation
- Search engines use semantic structure to understand content hierarchy
- Proper heading structure (`<h1>`–`<h6>`) is crucial—don't skip levels
- `<article>` and `<section>` are not interchangeable—consider if content is self-contained
- Always use `<main>` once per page for primary content
