## 4.1 Introduction to JavaScript

### What is JavaScript?

JavaScript is a **dynamic, interpreted programming language** that conforms to the ECMAScript specification. Unlike HTML and CSS, which are markup and style sheet languages respectively, JavaScript is a full-fledged programming language capable of complex calculations, data manipulation, and logical operations.

**Key characteristics of JavaScript:**

- **High-level**: Abstracts away complex computer science concepts
- **Dynamic**: Types are determined at runtime (flexible but requires care)
- **Interpreted**: Executed by the browser's JavaScript engine without prior compilation
- **Prototype-based**: Objects inherit directly from other objects
- **Multi-paradigm**: Supports procedural, object-oriented, and functional programming
- **Event-driven**: Designed to respond to user interactions and system events

### JavaScript vs Java

Despite the similar name, JavaScript and Java are completely different languages:

- **Java**: Statically typed, compiled, runs on JVM, used for Android/enterprise
- **JavaScript**: Dynamically typed, interpreted, runs in browsers/Node.js, used for web

The name "JavaScript" was a marketing decision to capitalize on Java's popularity in the 1990s. The official name is ECMAScript, though everyone calls it JavaScript.

## 4.2 Including JavaScript in HTML

There are three primary ways to include JavaScript in your HTML documents. Each has specific use cases and implications for performance and maintainability.

### Inline JavaScript

Inline JavaScript is written directly within HTML elements using event handler attributes:

```html
<!-- Inline JavaScript in event attributes -->
<button onclick="alert('Hello World!')">Click Me</button>

<a href="#" onmouseover="console.log('Hovered!')">Hover over me</a>

<form onsubmit="return validateForm()">
    <!-- form content -->
</form>
```

**Characteristics:**

| Aspect | Detail |
|--------|--------|
| **Scope** | Single element |
| **Maintainability** | Poor (mixed with HTML) |
| **Performance** | Slower (can't be cached separately) |
| **Security** | Risk of XSS (Cross-Site Scripting) |
| **Use case** | Quick prototypes only |

**Industry Standard:** **Avoid inline JavaScript** in production. It violates the principle of separation of concerns and creates security vulnerabilities.

### Internal (Embedded) JavaScript

Internal scripts are placed within `<script>` tags in the HTML document:

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Internal JavaScript</title>
    
    <!-- Script in head -->
    <script>
        // Code runs immediately when parsed
        console.log('Page is loading...');
    </script>
</head>
<body>
    <h1>Welcome</h1>
    
    <!-- Script at end of body -->
    <script>
        // Code runs after HTML is parsed
        console.log('Page loaded');
        
        // Select element and modify it
        const heading = document.querySelector('h1');
        heading.style.color = 'blue';
    </script>
</body>
</html>
```

### External JavaScript Files

External files are the **industry standard** for production websites:

**script.js:**
```javascript
// External JavaScript file: script.js

// Wait for DOM to be fully loaded
document.addEventListener('DOMContentLoaded', function() {
    console.log('External script loaded');
    
    // Initialize application
    initializeApp();
});

function initializeApp() {
    const button = document.querySelector('#cta-button');
    if (button) {
        button.addEventListener('click', handleClick);
    }
}

function handleClick(event) {
    console.log('Button clicked!', event);
    alert('Thanks for clicking!');
}
```

**index.html:**
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>External JavaScript</title>
    
    <!-- External script in head with defer -->
    <script src="script.js" defer></script>
</head>
<body>
    <h1>External JavaScript Example</h1>
    <button id="cta-button">Click Me</button>
</body>
</html>
```

## 4.3 Variables and Constants

Variables are containers for storing data values. JavaScript provides three ways to declare variables, each with different scoping rules.

### The Three Declaration Keywords

```javascript
// var - Function scoped, can be redeclared and updated (Legacy, avoid)
var name = 'John';
var name = 'Jane';  // Valid: Redeclaration allowed
name = 'Bob';       // Valid: Update allowed

// let - Block scoped, can be updated but not redeclared (Use for mutable data)
let age = 25;
age = 26;           // Valid: Update allowed
// let age = 30;    // Error: Cannot redeclare in same scope

// const - Block scoped, cannot be updated or redeclared (Use for immutable data)
const PI = 3.14159;
// PI = 3.14;       // Error: Cannot reassign constant
// const PI = 3;    // Error: Cannot redeclare
```

**Comparison Table:**

| Feature | `var` | `let` | `const` |
|---------|-------|-------|---------|
| Scope | Function | Block | Block |
| Hoisting | Yes (initialized undefined) | Yes (not initialized) | Yes (not initialized) |
| Redeclaration | Allowed | Not allowed | Not allowed |
| Reassignment | Allowed | Allowed | Not allowed |
| Temporal Dead Zone | No | Yes | Yes |
| Recommendation | Avoid | Use for changing values | Use for fixed values |

### Block vs Function Scope

Understanding scope is crucial for avoiding bugs:

```javascript
// var - Function scoped
function example() {
    if (true) {
        var message = 'Hello';
        console.log(message);  // 'Hello'
    }
    console.log(message);      // 'Hello' - still accessible (function scope)
}

// let/const - Block scoped
function example() {
    if (true) {
        let message = 'Hello';
        const PI = 3.14;
        console.log(message);  // 'Hello'
    }
    // console.log(message);   // Error: message is not defined
    // console.log(PI);        // Error: PI is not defined
}

// Global scope pollution
var globalVar = 'I am global';      // Creates property on window object
let globalLet = 'I am not global';  // Does not pollute global scope
```

### Temporal Dead Zone (TDZ)

Variables declared with `let` and `const` exist in a "temporal dead zone" from the start of the block until the declaration is encountered:

## 4.4 Data Types

JavaScript is a **dynamically typed** language, meaning you don't declare types explicitly, and variables can hold values of any type.

### Primitive Data Types

JavaScript has seven primitive data types:

```javascript
// 1. String - Textual data
const name = 'John Doe';
const greeting = "Hello World";
const template = `Hello, ${name}`;  // Template literal (ES6)

// 2. Number - Numeric values (integers and floats)
const age = 25;
const price = 19.99;
const negative = -10;
const infinity = Infinity;
const notANumber = NaN;  // Not a Number (result of invalid math)

// 3. BigInt - Large integers beyond Number.MAX_SAFE_INTEGER
const hugeNumber = 9007199254740991n;
const big = BigInt(12345678901234567890);

// 4. Boolean - Logical values
const isActive = true;
const hasPermission = false;

// 5. Undefined - Uninitialized variable
let uninitialized;
console.log(uninitialized);  // undefined

// 6. Null - Intentional absence of value
const emptyValue = null;

// 7. Symbol - Unique identifier (ES6)
const uniqueId = Symbol('description');
const anotherId = Symbol('description');
console.log(uniqueId === anotherId);  // false - always unique
```

### The typeof Operator

Use `typeof` to check the type of a value:

```javascript
console.log(typeof 'Hello');        // "string"
console.log(typeof 42);             // "number"
console.log(typeof 42n);            // "bigint"
console.log(typeof true);           // "boolean"
console.log(typeof undefined);      // "undefined"
console.log(typeof Symbol('x'));    // "symbol"
console.log(typeof {});             // "object" (all objects)
console.log(typeof []);             // "object" (arrays are objects)
console.log(typeof null);           // "object" (historical bug, should be null)
console.log(typeof function(){});   // "function" (functions are objects too)
```

**The `typeof null` bug:**
`typeof null` returns `"object"` due to a bug in the original JavaScript implementation that was never fixed for backward compatibility. To check for null:

```javascript
function isNull(value) {
    return value === null;
}
```

### Type Coercion and Conversion

JavaScript automatically converts types in certain operations (coercion), but you can also convert explicitly:

```javascript
// Implicit Coercion (Automatic)
console.log('5' + 3);       // "53" (number converted to string)
console.log('5' - 3);       // 2 (string converted to number)
console.log('5' * 2);       // 10 (string converted to number)
console.log(true + 1);      // 2 (true becomes 1)
console.log(false + 1);     // 1 (false becomes 0)
console.log(!!'hello');     // true (string to boolean)

// Explicit Conversion (Recommended - more readable)

// To String
String(123);                // "123"
(123).toString();           // "123"
String(true);               // "true"
String(null);               // "null"
String(undefined);          // "undefined"

// To Number
Number('123');              // 123
Number('123.45');           // 123.45
Number('');                 // 0
Number('hello');            // NaN
Number(true);               // 1
Number(false);              // 0
Number(null);               // 0
Number(undefined);          // NaN

// Alternative number conversion (unary plus)
const num = +'123';         // 123

// Parse integers and floats
parseInt('123px');          // 123 (stops at non-digit)
parseInt('101', 2);         // 5 (binary to decimal)
parseFloat('123.45.67');    // 123.45 (stops at second dot)

// To Boolean
Boolean(1);                 // true
Boolean(0);                 // false
Boolean('');                // false
Boolean('hello');           // true
Boolean(null);              // false
Boolean(undefined);         // false
Boolean({});                // true (objects are truthy)
Boolean([]);                // true (arrays are truthy)

// Double negation shortcut
!!'hello';                  // true
!!'';                       // false
!!0;                        // false
!!1;                        // true
```

### Truthy and Falsy Values

In boolean contexts (like `if` statements), values are coerced to `true` or `false`:

**Falsy values** (evaluate to `false`):
- `false`
- `0`
- `""` (empty string)
- `null`
- `undefined`
- `NaN`
- `0n` (BigInt zero)

**Truthy values** (everything else):
- `true`
- Any non-zero number
- Any non-empty string (including `"0"` and `"false"`)
- Any object (including `{}` and `[]`)
- `Infinity`
- Non-zero BigInt

```javascript
// Common pitfall with falsy values
if (0) { console.log('never runs'); }
if ('') { console.log('never runs'); }
if (null) { console.log('never runs'); }

// But be careful:
const userInput = '0';
if (userInput) {
    console.log('Runs! String "0" is truthy');
}

// Checking for existence:
const value = 0;
if (value) {
    // Won't run because 0 is falsy
}

// Better explicit checks:
if (value !== undefined && value !== null) {
    // Runs for 0, '', etc.
}

// Modern nullish check:
if (value != null) {  // catches both null and undefined
    // Runs for 0 and ''
}
```

## 4.5 Numbers and Math

JavaScript uses the IEEE 754 double-precision 64-bit binary format for numbers, allowing representation of both integers and floating-point numbers.

### Number Basics

```javascript
// Integers and floats
const integer = 42;
const float = 3.14159;
const negative = -10;

// Scientific notation
const billion = 1e9;        // 1000000000
const microsecond = 1e-6;   // 0.000001

// Hexadecimal, binary, octal
const hex = 0xff;           // 255
const binary = 0b11111111;  // 255
const octal = 0o377;        // 255

// Underscores as separators (ES2021)
const million = 1_000_000;  // 1000000
const creditCard = 1234_5678_9012_3456;

// Special numeric values
console.log(1 / 0);         // Infinity
console.log(-1 / 0);        // -Infinity
console.log('text' / 2);    // NaN (Not a Number)
```

### Arithmetic Operators

```javascript
// Basic operators
const sum = 5 + 3;          // Addition: 8
const difference = 5 - 3;   // Subtraction: 2
const product = 5 * 3;      // Multiplication: 15
const quotient = 5 / 3;     // Division: 1.666...
const remainder = 5 % 3;    // Modulo (remainder): 2
const power = 5 ** 3;       // Exponentiation: 125

// Increment and decrement
let counter = 0;
counter++;                  // Increment by 1 (returns old value)
++counter;                  // Increment by 1 (returns new value)
counter--;                  // Decrement by 1
--counter;                  // Decrement by 1

// Assignment operators
let x = 10;
x += 5;                     // x = x + 5 (15)
x -= 3;                     // x = x - 3 (12)
x *= 2;                     // x = x * 2 (24)
x /= 4;                     // x = x / 4 (6)
x %= 4;                     // x = x % 4 (2)
x **= 3;                    // x = x ** 3 (8)
```

**Increment/Decrement prefix vs postfix:**

```javascript
let a = 1;
let b = ++a;  // a becomes 2, b is 2 (increment happens first)

let c = 1;
let d = c++;  // d is 1, then c becomes 2 (assignment happens first)

console.log(a, b);  // 2, 2
console.log(c, d);  // 2, 1
```

### Floating Point Precision

JavaScript has quirks with floating-point arithmetic due to binary representation:

```javascript
// Classic floating point issue
console.log(0.1 + 0.2);  // 0.30000000000000004, not 0.3
console.log(0.1 + 0.2 === 0.3);  // false!

// Solutions:
// 1. Use integers (cents instead of dollars)
const price1 = 100;  // $1.00
const price2 = 200;  // $2.00
const total = (price1 + price2) / 100;  // 3.00

// 2. Round to fixed decimals
const sum = 0.1 + 0.2;
console.log(Math.round(sum * 100) / 100);  // 0.3
console.log(parseFloat(sum.toFixed(2)));   // 0.3

// 3. Use libraries for financial calculations (decimal.js, big.js)
```

### The Math Object

The built-in `Math` object provides mathematical constants and functions:

```javascript
// Constants
Math.PI;                    // 3.141592653589793
Math.E;                     // 2.718281828459045 (Euler's number)
Math.LN2;                   // 0.6931471805599453 (natural log of 2)
Math.LN10;                  // 2.302585092994046 (natural log of 10)
Math.SQRT2;                 // 1.4142135623730951 (square root of 2)

// Rounding
Math.round(3.7);            // 4 (round to nearest integer)
Math.round(3.2);            // 3
Math.ceil(3.2);             // 4 (round up)
Math.floor(3.7);            // 3 (round down)
Math.trunc(3.7);            // 3 (remove decimal part)

// Absolute value and sign
Math.abs(-5);               // 5
Math.sign(-5);              // -1 (negative)
Math.sign(5);               // 1 (positive)
Math.sign(0);               // 0

// Min and Max
Math.max(1, 5, 3, 9, 2);    // 9
Math.min(1, 5, 3, 9, 2);    // 1

// With arrays (spread operator)
const numbers = [1, 5, 3, 9, 2];
Math.max(...numbers);       // 9
Math.min(...numbers);       // 1

// Powers and roots
Math.pow(2, 3);             // 8 (2^3)
Math.sqrt(16);              // 4 (square root)
Math.cbrt(27);              // 3 (cube root)

// Random numbers
Math.random();              // Random number between 0 (inclusive) and 1 (exclusive)

// Random integer between min and max (inclusive)
function getRandomInt(min, max) {
    return Math.floor(Math.random() * (max - min + 1)) + min;
}

getRandomInt(1, 10);        // Random integer from 1 to 10

// Trigonometry
Math.sin(Math.PI / 2);      // 1 (sine)
Math.cos(Math.PI);          // -1 (cosine)
Math.tan(0);                // 0 (tangent)

// Logarithms
Math.log(Math.E);           // 1 (natural log)
Math.log10(100);            // 2 (base 10 log)
Math.log2(8);               // 3 (base 2 log)
```

### Number Methods

```javascript
const num = 123.456789;

// toFixed - fixed decimal places (returns string)
num.toFixed(2);             // "123.46" (rounded)
num.toFixed(0);             // "123"

// toPrecision - total significant digits
num.toPrecision(4);         // "123.5"
(1234.5).toPrecision(2);    // "1.2e+3" (scientific notation)

// toString - convert to string with base
(255).toString();           // "255"
(255).toString(16);         // "ff" (hexadecimal)
(255).toString(2);          // "11111111" (binary)
(255).toString(8);          // "377" (octal)

// Checking numbers
Number.isFinite(42);        // true
Number.isFinite(Infinity);  // false
Number.isFinite('42');      // false (doesn't coerce)

Number.isInteger(42);       // true
Number.isInteger(42.5);     // false

Number.isNaN(NaN);          // true
Number.isNaN('NaN');        // false (doesn't coerce)
isNaN('NaN');               // true (coerces first - avoid)

// Parsing
Number.parseInt('100px');   // 100
Number.parseFloat('12.5em'); // 12.5

// Exponential notation
(1234567).toExponential();  // "1.234567e+6"
(1234567).toExponential(2); // "1.23e+6"
```

## 4.6 Strings

Strings are sequences of characters used to represent text. JavaScript strings are immutable (cannot be changed once created).

### Creating Strings

```javascript
// Single quotes
const name = 'John';

// Double quotes
const greeting = "Hello";

// Backticks (template literals) - ES6
const message = `Hello, ${name}!`;

// Multi-line with backticks
const poem = `
Roses are red,
Violets are blue,
JavaScript is awesome,
And so are you!
`;

// Escape characters
const quote = "He said, \"Hello!\"";  // Escaped quotes
const newline = "Line 1\nLine 2";     // New line
const tab = "Column 1\tColumn 2";     // Tab
const backslash = "C:\\Users\\Name";  // Backslash

// Accessing characters (strings are array-like)
const text = "Hello";
console.log(text[0]);       // "H"
console.log(text.charAt(0)); // "H"
console.log(text[text.length - 1]); // "o" (last character)

// Strings are immutable
text[0] = 'h';              // No error, but doesn't work
console.log(text);          // Still "Hello"
```

### String Methods

**Changing case:**
```javascript
"Hello".toLowerCase();      // "hello"
"Hello".toUpperCase();      // "HELLO"
```

**Searching:**
```javascript
const text = "The quick brown fox jumps over the lazy dog";

// indexOf - first occurrence
text.indexOf("quick");      // 4 (position)
text.indexOf("cat");        // -1 (not found)
text.indexOf("o");          // 12 (first 'o')
text.indexOf("o", 13);      // 17 (search from position 13)

// lastIndexOf - last occurrence
text.lastIndexOf("o");      // 41 (last 'o')

// includes - check existence (ES6)
text.includes("quick");     // true
text.includes("cat");       // false
text.includes("quick", 5);  // false (starts searching from index 5)

// startsWith and endsWith (ES6)
text.startsWith("The");     // true
text.startsWith("quick", 4); // true (starts at position 4)
text.endsWith("dog");       // true
text.endsWith("lazy", 40);  // true (first 40 characters end with "lazy")
```

**Extracting substrings:**
```javascript
const str = "Hello, World!";

// slice(start, end) - end not included
str.slice(0, 5);            // "Hello"
str.slice(7);               // "World!" (from 7 to end)
str.slice(-6);              // "World!" (from end)
str.slice(-6, -1);          // "World" (negative indices)

// substring(start, end) - similar but doesn't accept negatives
str.substring(0, 5);        // "Hello"
str.substring(7, 12);       // "World"

// substr(start, length) - legacy, avoid if possible
str.substr(7, 5);           // "World"

// split - convert to array
str.split(" ");             // ["Hello,", "World!"]
str.split("");              // ["H", "e", "l", "l", "o", ...]
str.split(", ");            // ["Hello", "World!"]
```

**Trimming and padding:**
```javascript
const padded = "  Hello World  ";

// trim - remove whitespace from both ends
padded.trim();              // "Hello World"
padded.trimStart();         // "Hello World  " (left only)
padded.trimEnd();           // "  Hello World" (right only)

// padStart and padEnd (ES6)
"5".padStart(2, "0");       // "05" (padding for numbers)
"5".padStart(3, "0");       // "005"
"Hello".padEnd(10, ".");    // "Hello....."
```

**Replacing:**
```javascript
const text = "The quick brown fox";

// replace - first occurrence only
text.replace("quick", "slow");  // "The slow brown fox"

// replace with regex - all occurrences (with g flag)
text.replace(/o/g, "0");        // "The quick br0wn f0x"

// replaceAll (ES2021) - all occurrences without regex
text.replaceAll("o", "0");      // "The quick br0wn f0x"

// Callback function for complex replacements
text.replace(/\b\w/g, char => char.toUpperCase());  // "The Quick Brown Fox"

// Important: replace returns new string (strings are immutable)
let newText = text.replace("quick", "slow");
console.log(text);              // Still "The quick brown fox"
console.log(newText);           // "The slow brown fox"
```

**Repeating:**
```javascript
"Hello".repeat(3);          // "HelloHelloHello"
"Na".repeat(2) + " Batman!"; // "NaNa Batman!"
"*".repeat(10);             // "**********"
```

### Template Literals (ES6)

Template literals provide powerful string interpolation and multi-line support:

```javascript
const name = "Alice";
const age = 30;

// Interpolation with ${}
const greeting = `Hello, ${name}! You are ${age} years old.`;

// Expressions inside ${}
const price = 10;
const tax = 0.2;
const total = `Total: $${(price * (1 + tax)).toFixed(2)}`;  // "Total: $12.00"

// Multi-line without \n
const html = `
    <div class="card">
        <h2>${name}</h2>
        <p>Age: ${age}</p>
    </div>
`;

// Tagged templates (advanced)
function highlight(strings, ...values) {
    return strings.reduce((result, string, i) => {
        const value = values[i] ? `**${values[i]}**` : '';
        return result + string + value;
    }, '');
}

const result = highlight`Hello ${name}, you are ${age} years old`;
// "Hello **Alice**, you are **30** years old"
```

## 4.7 Booleans and Comparisons

Boolean logic is fundamental to programming. JavaScript provides comparison operators and logical operators to evaluate conditions.

### Comparison Operators

```javascript
// Equality
5 == 5;                     // true (loose equality)
5 == '5';                   // true (coerces string to number)
5 === 5;                    // true (strict equality - recommended)
5 === '5';                  // false (different types)

// Inequality
5 != '5';                   // false (coerces, then compares)
5 !== '5';                  // true (strict - different types)

// Relational
5 > 3;                      // true
5 < 3;                      // false
5 >= 5;                     // true
5 <= 4;                     // false

// Comparing strings (lexicographic)
'apple' < 'banana';         // true
'Apple' < 'apple';          // true (uppercase < lowercase)
```

**Strict equality (===) vs Loose equality (==):**

```javascript
// Loose equality (==) performs type coercion
0 == false;                 // true (0 coerced to false)
0 == '0';                   // true (string coerced to number)
'' == false;                // true (empty string coerced to 0)
null == undefined;          // true (special case)

// Strict equality (===) requires same type and value
0 === false;                // false
0 === '0';                  // false
null === undefined;         // false

// Industry Standard: Always use === and !== to avoid unexpected coercion
```

### Logical Operators

```javascript
// AND (&&) - both must be true
true && true;               // true
true && false;              // false

// OR (||) - at least one must be true
true || false;              // true
false || false;             // false

// NOT (!) - inverts
!true;                      // false
!false;                     // true

// Double negation (convert to boolean)
!!'hello';                  // true
!!'';                       // false
!!0;                        // false
```

**Short-circuit evaluation:**

```javascript
// AND (&&) returns first falsy value or last value
console.log(true && 'hello');       // 'hello'
console.log(false && 'hello');      // false
console.log(1 && 2 && 3);           // 3 (all truthy, returns last)
console.log(1 && 0 && 3);           // 0 (returns first falsy)

// OR (||) returns first truthy value or last value
console.log(true || 'hello');       // true
console.log(false || 'hello');      // 'hello'
console.log(0 || '' || null);       // null (all falsy, returns last)
console.log(0 || 'default');        // 'default'

// Practical use: Default values (before ?? existed)
const username = input || 'Anonymous';

// Nullish coalescing (??) - ES2020
// Returns right side only if left is null or undefined (not 0 or '')
const count = 0 ?? 10;              // 0 (0 is not nullish)
const name = null ?? 'Anonymous';   // 'Anonymous'
```

### Common Comparison Pitfalls

```javascript
// NaN is not equal to anything, including itself
NaN === NaN;                // false
isNaN(NaN);                 // true (check with function)
Number.isNaN(NaN);          // true (modern check)

// Object comparison compares references, not values
const obj1 = { a: 1 };
const obj2 = { a: 1 };
const obj3 = obj1;

obj1 === obj2;              // false (different objects in memory)
obj1 === obj3;              // true (same reference)

// Array comparison
[1, 2] === [1, 2];          // false (different objects)
const arr = [1, 2];
arr === arr;                // true (same reference)

// Comparing objects requires deep comparison function
function deepEqual(obj1, obj2) {
    return JSON.stringify(obj1) === JSON.stringify(obj2);
    // Note: Simple method, has limitations with undefined, functions, etc.
}
```
