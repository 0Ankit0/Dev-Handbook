## **46.1 Introduction to Scrum**

Scrum is the most widely adopted Agile framework, designed to deliver complex products iteratively and incrementally. It emphasizes collaboration, accountability, and iterative progress toward a well-defined goal. For testers, understanding Scrum is essential because testing activities are tightly interwoven with Scrum events, artifacts, and roles.

### **46.1.1 Scrum Framework Overview**

Scrum is built on three pillars: **Transparency**, **Inspection**, and **Adaptation**. It comprises:

- **Roles:** Product Owner, Scrum Master, Developers (including testers)
- **Events:** Sprint, Sprint Planning, Daily Scrum, Sprint Review, Sprint Retrospective
- **Artifacts:** Product Backlog, Sprint Backlog, Increment

### **46.1.2 Key Scrum Roles and Testing**

| Role | Testing Responsibilities |
|------|--------------------------|
| **Product Owner** | Defines acceptance criteria; collaborates with testers to clarify requirements; prioritizes defect fixes |
| **Scrum Master** | Coaches team on Scrum and testing best practices; removes impediments to testing (environments, tools) |
| **Developers (incl. Testers)** | Design, develop, and test the product; write and automate tests; ensure quality of the Increment |

In Scrum, testers are considered part of the **Development Team** (now called "Developers" in the 2020 Scrum Guide). There is no separate "tester" role; everyone shares responsibility for quality. However, in practice, team members may specialize in testing while still contributing to development tasks.

## **46.2 Testing in Scrum Events**

### **46.2.1 Sprint Planning**

Sprint Planning answers two questions: **What** can be delivered this sprint? and **How** will the work be done? Testing is integral to both.

**Testing activities during Sprint Planning:**

- **Reviewing selected Product Backlog items** for testability, clarity, and completeness of acceptance criteria.
- **Collaborating with the Product Owner** to refine acceptance criteria, adding edge cases and examples (using techniques like Example Mapping).
- **Estimating testing effort** for each story, ensuring the team accounts for test design, automation, and execution.
- **Defining the testing strategy** for the sprint: what will be automated, what will be explored manually, and what test data/environments are needed.
- **Identifying dependencies** (e.g., APIs, test data) that may impact testing.

**Output:** A Sprint Backlog that includes testing tasks (e.g., "Write automated acceptance tests for login", "Perform exploratory testing on checkout").

### **46.2.2 Daily Scrum (Stand-up)**

The Daily Scrum is a 15-minute event for Developers to inspect progress and adapt the plan. Testers participate alongside developers.

**Testing-focused contributions:**

- **Report progress** on testing tasks (e.g., "I automated three test cases for story #42").
- **Raise blockers** affecting testing (e.g., "The test environment is down", "I need test data from the API team").
- **Coordinate with developers** on fixes and retesting (e.g., "The bug I reported yesterday is fixed; I'll retest it today").
- **Share insights** from exploratory testing that may impact the sprint goal.

**Important:** The Daily Scrum is not a status meeting for the Scrum Master; it's for Developers to self-organize. Testers should speak about their work and commit to helping the team meet the sprint goal.

### **46.2.3 Sprint Review**

The Sprint Review is held at the end of the sprint to inspect the Increment and adapt the Product Backlog. Stakeholders provide feedback.

**Testing involvement in Sprint Review:**

- **Demonstrate** the working product, highlighting features and their quality.
- **Present test results** (e.g., automated test pass rates, exploratory testing findings) to build confidence.
- **Discuss defects** found and fixed, as well as any known issues.
- **Gather feedback** from stakeholders on usability, performance, or missing scenarios, which may become new backlog items.
- **Showcase test automation** improvements (e.g., faster test execution, new test coverage) as part of the team's achievements.

### **46.2.4 Sprint Retrospective**

The Sprint Retrospective is an opportunity for the team to inspect itself and create a plan for improvements. Testing practices are a common topic.

**Testing-related discussions:**

- **What went well** in testing? (e.g., early bug detection, effective automation)
- **What didn't go well?** (e.g., flaky tests, insufficient test data)
- **What can we improve?** (e.g., add more unit tests, invest in test environment stability, pair testing sessions)
- **Action items** for the next sprint (e.g., "Refactor the fragile login tests", "Create a shared test data pool").

**Example retrospective outcome:**
> "We noticed that our UI tests are slow and flaky. Next sprint, we'll dedicate two story points to refactoring them using a better waiting strategy and reducing duplication."

## **46.3 Definition of Ready (DoR) and Definition of Done (DoD)**

In Scrum, teams often define **Ready** and **Done** criteria to ensure consistent understanding of when work can start and when it's complete.

### **46.3.1 Definition of Ready (DoR)**

DoR defines when a Product Backlog item is ready to be taken into a sprint. From a testing perspective:

**Typical DoR criteria:**
- User story is clearly described with a well-defined **Acceptance Criteria** (Given-When-Then format preferred).
- Acceptance criteria are **testable** (no ambiguous statements).
- Dependencies (e.g., external APIs, test data) are identified and available or mocked.
- The story has been estimated by the team, and testing effort is included.
- Non-functional requirements (performance, security) are defined if applicable.
- The Product Owner has confirmed the story's value and priority.

**Why DoR matters for testers:** If a story enters the sprint without clear, testable criteria, testers waste time clarifying requirements later, leading to delays and defects. A strong DoR prevents this.

### **46.3.2 Definition of Done (DoD)**

DoD is a shared understanding of what it means for work to be complete. It ensures consistency and transparency. Testing criteria are a core part of DoD.

**Typical DoD including testing:**
- All acceptance criteria are met and verified (automated where possible).
- Unit tests are written and passing.
- Integration tests (API, service) pass.
- Regression tests for affected areas pass.
- Code has been reviewed (including test code).
- Documentation (if any) is updated.
- Non-functional tests (performance, security) have been executed if applicable.
- No critical or high-severity defects remain open.
- The feature is deployed to a staging environment and verified.
- Exploratory testing has been performed on the feature.

**Importance:** A clear DoD ensures the team has a shared standard of quality. Without it, some team members may consider a story "done" before it's truly ready for release.

### **46.3.3 Evolving DoR and DoD**

DoR and DoD are not static; they should be reviewed and adapted in retrospectives. For example, a team might add "Automated acceptance tests are committed" to DoD after noticing manual regression testing takes too long.

## **46.4 Tester's Role in Scrum Team**

In Scrum, testers are Developers with testing expertise. They contribute to all aspects of the sprint.

| Activity | Tester's Role |
|----------|---------------|
| **Refinement** | Ask questions, write examples, ensure testability, estimate testing effort. |
| **Development** | Pair with developers on test automation; write unit/integration tests; code review with testing perspective. |
| **Testing** | Execute automated and manual tests; perform exploratory testing; log defects; verify fixes. |
| **Collaboration** | Communicate with Product Owner on acceptance criteria; help developers reproduce bugs; share test results. |
| **Continuous Improvement** | Participate in retrospectives; suggest improvements to testing process, tools, and automation. |

**Key Mindset Shift:** Testers in Scrum are not "handoffs" who test after development; they are partners who help build quality in from the start.

## **46.5 Testing Artifacts in Scrum**

Scrum defines minimal artifacts, but teams often create additional ones to support testing.

### **46.5.1 Sprint Backlog with Testing Tasks**

The Sprint Backlog includes testing tasks, which can be tracked on a physical or digital board (e.g., Jira, Trello).

Example tasks:
- "Write unit tests for UserService"
- "Automate acceptance test for login"
- "Perform exploratory testing on checkout flow"
- "Set up test data for discount codes"
- "Run performance test on product search"

### **46.5.2 Test Reports**

While not a formal Scrum artifact, teams often produce test reports to communicate quality:

- **Defect metrics** (open/closed, severity) â€“ visible to the team.

### **46.5.3 Burndown/Burnup Charts**

Burndown charts track remaining work. Testing tasks are included in the remaining work, so burndown reflects testing progress. If testing tasks are not completed, the chart will show it, prompting the team to adjust.

## **46.6 Common Challenges and Solutions**

| Challenge | Solution |
|-----------|----------|
| **Testing squeezed at end of sprint** | Shift left: test earlier, automate, involve testers in development, use TDD/BDD. |
| **Unclear acceptance criteria** | Collaborate with PO during refinement; use Example Mapping; write AC in Gherkin. |
| **Flaky automated tests** | Investigate root causes; improve test design; isolate from external dependencies; use retries sparingly. |
| **Insufficient test environments** | Use containerization (Docker) to spin up ephemeral environments; invest in infrastructure as code. |
| **Poor communication with developers** | Sit together, use pair testing, participate in daily stand-ups, use chat tools effectively. |
| **Definition of Done not respected** | Make DoD visible; include it in code review checklist; hold team accountable in retrospectives. |
| **Test data management issues** | Create scripts to generate test data; use dedicated test data sets; clean up after tests. |

## **46.7 Case Study: Testing in a Scrum Sprint**

**Team:** 7 developers (2 with testing focus), Product Owner, Scrum Master.  
**Project:** E-commerce mobile app.  
**Sprint Goal:** Enable customers to apply discount codes during checkout.

### **Sprint Planning (Day 1)**

- Team selects 4 stories related to discount codes.
- For story "Apply valid discount code", testers ask: "What happens if code is expired? Case sensitive? Multiple uses?" PO clarifies, and acceptance criteria are updated.
- Team estimates total effort: 20 story points (including 5 for testing tasks: automation, exploratory testing, test data setup).
- Testing tasks are added to Sprint Backlog.

### **During Sprint (Days 2-9)**

- **Day 2:** Developer starts coding; tester writes unit tests for discount calculation logic (TDD).
- **Day 4:** Feature ready in dev environment. Tester runs initial exploratory tests, finds bug: applying code after order submission still shows success but doesn't apply discount. Logs defect.
- **Day 5:** Developer fixes bug; tester retests and verifies.
- **Daily stand-ups:** Testers report progress, no blockers.
- **Day 7:** Automated acceptance tests are written and integrated into CI pipeline.
- **Day 8:** Tester performs exploratory testing on edge cases (expired codes, maximum usage, combining with other promotions). Finds minor usability issue (error message unclear). Logs as low priority; team agrees to fix next sprint.

### **Sprint Review (Day 10)**

- Team demonstrates applying discount codes with various scenarios.
- Tester presents test results: automated tests pass, exploratory testing revealed one minor issue (logged).
- Stakeholders are satisfied; provide feedback on wanting more types of promotions (e.g., BOGO) for future.

### **Sprint Retrospective (Day 10)**

- **What went well:** Early bug detection, good collaboration between dev and tester, automated tests ready.
- **What could improve:** Test data for expired codes was hard to set up; took time.
- **Action item:** Next sprint, create a script to generate test coupons with varied expiry dates.

## **Chapter Summary**

In this chapter, we explored how testing integrates with the Scrum framework:

- **Scrum events** (Sprint Planning, Daily Scrum, Sprint Review, Sprint Retrospective) all have testing activities that contribute to transparency, inspection, and adaptation.
- **Definition of Ready (DoR)** ensures stories are testable before entering a sprint, reducing waste.
- **Definition of Done (DoD)** includes testing criteria that guarantee quality and consistency.
- **Testers in Scrum** are Developers with testing expertise, collaborating throughout the sprint, not just at the end.
- **Testing artifacts** like sprint backlog tasks, test reports, and burndown charts support progress tracking.
- **Common challenges** have practical solutions, from shifting left to improving test data management.

**Key Insight:** Scrum provides a rhythm of events and artifacts that naturally incorporate testing when the team embraces quality as a shared responsibility. Testers are not separate; they are integral to the team's success.
