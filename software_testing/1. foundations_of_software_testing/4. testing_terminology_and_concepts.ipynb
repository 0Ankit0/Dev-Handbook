## **4.1 Key Testing Vocabulary**

Before diving into testing practices, you must speak the language of quality assurance. This section establishes the foundational terminology used across the industry, aligned with **IEEE 610.12** (Standard Glossary of Software Engineering Terminology) and **ISTQB** (International Software Testing Qualifications Board) standards.

### **4.1.1 Fundamental Testing Concepts**

#### **Testing vs. Checking**
While often used interchangeably, these have distinct meanings:

- **Testing:** The process of evaluating a system to learn about its quality, including exploration, experimentation, and learning. It involves human judgment and creativity.
- **Checking:** The process of confirming that specific expected behaviors occur. This can be automated.

```python
# Example: Checking vs Testing

class CheckingExample:
    """
    Checking: Verifying specific known expectations
    Can be automated
    """
    def check_login_with_valid_credentials(self):
        expected_result = "Dashboard page displayed"
        actual_result = login("user@example.com", "valid_pass")
        
        assert actual_result == expected_result, "Login check failed"
        return True  # Binary result: Pass/Fail

class TestingExample:
    """
    Testing: Exploring to discover information
    Requires human intelligence and exploration
    """
    def explore_login_security(self):
        """
        Tester explores various scenarios creatively:
        - What happens with Unicode characters in password?
        - What happens if I copy-paste into password field?
        - What happens if I manipulate the request in Burp Suite?
        - What happens after 1000 failed attempts?
        """
        observations = []
        
        # Exploratory observation 1
        result = login("admin'--", "anything")
        if "database error" in result:
            observations.append("SQL Injection vulnerability detected")
        
        # Exploratory observation 2
        start_time = time.time()
        login("user", "wrong")
        end_time = time.time()
        if end_time - start_time > 5:
            observations.append("Potential timing attack vector")
        
        # Testing produces information, not just pass/fail
        return observations
```

#### **Verification vs. Validation**

These are often confused but address different questions:

| Aspect | Verification | Validation |
|--------|-------------|------------|
| **Question** | "Are we building it right?" | "Are we building the right thing?" |
| **Focus** | Process, documentation, design | Product, user needs, requirements |
| **Timing** | Throughout development | Usually end of development |
| **Methods** | Reviews, inspections, walkthroughs | Testing, user acceptance, beta |
| **IEEE Definition** | Confirmation that requirements are met | Confirmation that user needs are met |

```python
# Examples distinguishing Verification and Validation

class VerificationActivities:
    """
    Static methods - examining artifacts without executing code
    """
    def verify_requirements_document(self):
        """Checking SRS for completeness and consistency"""
        checks = {
            "traceability": "Each requirement has unique ID",
            "testability": "Requirements are measurable",
            "consistency": "No contradictions between requirements",
            "completeness": "All user stories have acceptance criteria"
        }
        return checks
    
    def verify_design_document(self):
        """Reviewing HLD/LLD against standards"""
        return {
            "architecture_review": "Design patterns properly applied",
            "interface_specs": "API contracts clearly defined",
            "database_schema": "Normalization rules followed"
        }

class ValidationActivities:
    """
    Dynamic methods - executing the software
    """
    def validate_user_requirements(self):
        """Confirming software meets actual user needs"""
        test_results = {
            "user_story_1": "User can complete checkout in 3 clicks",
            "user_story_2": "Search returns relevant results in <2 seconds",
            "acceptance_criteria": "All business rules enforced"
        }
        return test_results
    
    def validate_business_value(self):
        """Confirming the software solves the business problem"""
        metrics = {
            "user_adoption_rate": "85% of beta users completed onboarding",
            "error_reduction": "Data entry errors reduced by 90%",
            "efficiency_gain": "Process time reduced from 2 hours to 15 minutes"
        }
        return metrics
```

#### **Error, Defect, Bug, Failure - The Testing Taxonomy**

Understanding the causal chain is crucial for effective defect management:

```
Human Error (Mistake) 
        ↓
Defect (Bug/Fault) - exists in code/document
        ↓
Failure - external behavior/result (what user sees)

Example:
[Error] Developer misunderstands requirement → 
[Defect] Code checks password length < 6 instead of >= 6 → 
[Failure] User with valid 8-character password cannot login
```

- **Error (Mistake):** Human action that produces an incorrect result (e.g., developer typo, misunderstood requirement).
- **Defect (Fault/Bug):** Imperfection in a work product where it does not meet its requirements or specifications (exists in code).
- **Failure:** Deviation of the software from its expected delivery or service (external behavior visible to users).
- **Incident:** Any event occurring that requires investigation (might be a failure, or might be expected behavior misunderstood by user).

```python
class DefectTaxonomy:
    """
    Understanding the lifecycle of a software problem
    """
    
    def analyze_chain_of_events(self):
        """
        Trace from human error to system failure
        """
        chain = {
            "human_error": {
                "type": "Commission",
                "description": "Developer implemented caching but forgot to invalidate cache on update",
                "cause": "Oversight during code review"
            },
            "defect": {
                "location": "user_service.py:142",
                "type": "Logic Error",
                "description": "Cache invalidation missing in update_user() method",
                "severity": "High"
            },
            "failure": {
                "symptom": "User updates profile but sees old data",
                "reproducibility": "Always, after cache TTL expires it fixes temporarily",
                "impact": "Data inconsistency, user confusion"
            }
        }
        return chain
    
    def classify_defect_type(self, defect_description):
        """
        Industry standard defect classifications
        """
        classifications = {
            "functional": "Software does not perform required function",
            "performance": "Response time, throughput, or resource usage unacceptable",
            "usability": "User interface confusing or non-intuitive",
            "reliability": "System crashes or hangs under load",
            "security": "Unauthorized access possible or data exposed",
            "compatibility": "Works on one platform but not another",
            "documentation": "Help text or manuals incorrect"
        }
        
        # Classification logic
        if "timeout" in defect_description or "slow" in defect_description:
            return "performance"
        elif "crash" in defect_description or "hang" in defect_description:
            return "reliability"
        # ... more logic
        
        return "functional"
```

#### **Test Condition**
An item or event that could be verified by one or more test cases. Think of these as "things to test."

**Example:**
- Test Condition: "User authentication"
- Test Cases: Valid login, invalid password, locked account, expired password, etc.

#### **Test Scenario**
A high-level description of a test case or set of test cases. It answers "What to test?" at a business level.

```
Test Scenario: Verify the login functionality
  ├── Test Case 1: Valid credentials
  ├── Test Case 2: Invalid password  
  ├── Test Case 3: Empty username
  └── Test Case 4: SQL injection attempt
```

#### **Test Case**
A set of preconditions, inputs, actions, expected results, and postconditions developed based on test conditions.

**Anatomy of a Test Case (IEEE 829 Standard):**
- **Test Case ID:** Unique identifier (e.g., TC_LOGIN_001)
- **Title/Description:** What is being tested
- **Preconditions:** State before test execution
- **Test Steps:** Detailed actions
- **Test Data:** Specific inputs
- **Expected Result:** Predicted outcome (Test Oracle)
- **Actual Result:** Observed outcome (filled during execution)
- **Status:** Pass/Fail/Blocked/Not Run
- **Priority:** Critical/High/Medium/Low

```python
class TestCase:
    """
    IEEE 829 compliant test case structure
    """
    def __init__(self):
        self.template = {
            "test_case_id": "TC_[Module]_[Number]",
            "title": "Brief description of what is tested",
            "module": "Component being tested",
            "author": "Test designer name",
            "creation_date": "YYYY-MM-DD",
            "priority": "Critical/High/Medium/Low",
            "type": "Positive/Negative/Regression",
            
            "preconditions": [
                "User must be registered",
                "Database must be populated",
                "Network connectivity available"
            ],
            
            "test_steps": [
                {"step_no": 1, "action": "Navigate to login page", "expected": "Login page loads"},
                {"step_no": 2, "action": "Enter valid username", "expected": "Username accepted"},
                {"step_no": 3, "action": "Enter valid password", "expected": "Password masked"},
                {"step_no": 4, "action": "Click Login button", "expected": "Dashboard displayed"}
            ],
            
            "test_data": {
                "username": "testuser@example.com",
                "password": "SecurePass123!"
            },
            
            "expected_result": "User successfully logs in and sees dashboard",
            "actual_result": "",  # To be filled during execution
            "status": "Not Executed",  # Pass/Fail/Blocked
            "defect_id": "",  # If failed, link to bug report
            "execution_date": "",
            "executed_by": "",
            "attachments": []  # Screenshots, logs
        }
    
    def example_login_test_case(self):
        """Concrete example"""
        return {
            "test_case_id": "TC_AUTH_001",
            "title": "Verify successful login with valid credentials",
            "module": "Authentication",
            "priority": "Critical",
            "preconditions": [
                "User 'john.doe' exists in database",
                "User account is active",
                "User is not locked out"
            ],
            "steps": [
                "Navigate to https://app.example.com/login",
                "Enter username: john.doe@example.com",
                "Enter password: CorrectPass123",
                "Click 'Sign In' button"
            ],
            "expected_result": "User redirected to /dashboard with welcome message 'Hello, John'",
            "test_data": {
                "username": "john.doe@example.com",
                "password": "CorrectPass123"
            }
        }
```

#### **Test Oracle**
A mechanism for determining whether a test has passed or failed. It's the source of expected results.

**Types of Test Oracles:**
1. **Specification-based:** Requirements documents
2. **Competitor-based:** Behavior of similar existing systems
3. **Heuristic:** Rules of thumb (e.g., "page should load in <3 seconds")
4. **Model-based:** Mathematical models predicting behavior
5. **Human:** Expert knowledge

```python
class TestOracle:
    """
    Determining expected results
    """
    
    def specification_oracle(self, requirement_id):
        """
        Expected result comes from requirements doc
        """
        requirements_db = {
            "REQ_001": "System shall calculate tax as 8% of subtotal",
            "REQ_002": "Password must be minimum 8 characters with 1 special character"
        }
        return requirements_db.get(requirement_id)
    
    def heuristic_oracle(self, observation):
        """
        Rules of thumb for expected behavior
        """
        heuristics = {
            "performance": lambda x: x < 3.0,  # Should be < 3 seconds
            "availability": lambda x: x >= 99.9,  # Should be 99.9% uptime
            "error_messages": lambda x: "user-friendly" in x  # Should be user-friendly
        }
        return heuristics
    
    def competitor_oracle(self, feature):
        """
        Compare against industry standard (competitor)
        """
        # Example: Our search should be as fast as Google's
        competitor_benchmarks = {
            "search_response_time": "0.5 seconds",
            "checkout_steps": 3,
            "form_validation": "real-time"
        }
        return competitor_benchmarks.get(feature)
```

## **4.2 Test Artifacts**

Test artifacts are the documents and deliverables produced during testing. Following **IEEE 829-2008** (Standard for Software Test Documentation), these ensure traceability, repeatability, and communication.

### **4.2.1 Master Test Plan (MTP)**

**Key Sections (IEEE 829):**
1. **Introduction:** Purpose, scope, references
2. **Test Items:** Software features to be tested
3. **Features Not to Be Tested:** Explicitly excluded scope
4. **Approach:** Testing strategy and techniques
5. **Item Pass/Fail Criteria:** When to accept/reject
6. **Suspension/Resumption Criteria:** When to pause/restart
7. **Test Deliverables:** Documents to be produced
8. **Testing Tasks:** Activities and dependencies
9. **Environmental Needs:** Hardware, software, tools
10. **Responsibilities:** Roles and assignments
11. **Staffing and Training Needs**
12. **Schedule:** Timeline and milestones
13. **Risks and Contingencies**
14. **Approvals:** Sign-off authorities

## **4.3 Defect/Bug Lifecycle**

The defect lifecycle defines the stages a bug goes through from discovery to closure. Understanding this workflow is essential for effective defect management and communication.

### **4.3.1 Defect States and Transitions**

**Standard Defect States (IEEE 1044):**

```
New → Assigned → Open → Fixed → Retest → 
  ↑                                      |
  └──────── Reopen ←──── Closed ←────────┘
  
Additional states:
- Rejected (not a bug)
- Duplicate (already reported)
- Deferred (fix later)
- Cannot Reproduce
- Verified
```

**State Definitions:**

1. **New:** Defect reported, not yet validated
2. **Assigned:** Allocated to a developer for investigation
3. **Open:** Developer confirms it's a bug and is working on fix
4. **Fixed:** Developer implemented fix, ready for testing
5. **Retest:** Tester verifying the fix
6. **Verified:** Fix confirmed working
7. **Closed:** Defect lifecycle complete
8. **Reopen:** Fix failed verification, back to developer
9. **Rejected:** Not a valid defect (works as designed)
10. **Duplicate:** Same issue already reported elsewhere
11. **Deferred:** Valid bug, but fix postponed to future release

```python
class DefectLifecycle:
    """
    Defect workflow management
    """
    
    def valid_transitions(self):
        """
        Valid state transitions in defect workflow
        """
        transitions = {
            "New": ["Assigned", "Rejected", "Duplicate", "Deferred"],
            "Assigned": ["Open", "Rejected", "Duplicate", "Deferred"],
            "Open": ["Fixed", "Cannot Reproduce", "Deferred", "Rejected"],
            "Fixed": ["Retest"],
            "Retest": ["Verified", "Reopen"],
            "Verified": ["Closed"],
            "Reopen": ["Assigned", "Fixed", "Rejected"],
            "Rejected": ["Closed", "New"],  # Can reopen if rejected in error
            "Duplicate": ["Closed"],
            "Deferred": ["Assigned", "Closed"],  # Can reopen later
            "Cannot Reproduce": ["Closed", "Assigned"]  # Tester can provide more info
        }
        return transitions
    
    def workflow_rules(self):
        """
        Business rules for defect workflow
        """
        rules = {
            "New_to_Assigned": "Must have valid component and severity",
            "Assigned_to_Open": "Developer acknowledges and accepts",
            "Open_to_Fixed": "Code committed and build available",
            "Fixed_to_Retest": "Build deployed to test environment",
            "Retest_to_Verified": "Test case passes in fix build",
            "Retest_to_Reopen": "Failure reproduced in fix build with evidence",
            "Any_to_Rejected": "Requires QA Lead approval and justification",
            "Any_to_Deferred": "Requires Product Owner and Tech Lead approval",
            "High_Critical_cannot_defer": "Severity 1-2 cannot be deferred without VP approval"
        }
        return rules
```

### **4.3.3 Defect Metrics and Analysis**

```python
class DefectMetrics:
    """
    Calculating defect-related metrics
    """
    
    def defect_detection_percentage(self, defects_found_pre_release, defects_found_post_release):
        """
        DDP = Defects found before release / Total defects found
        
        Higher is better (ideally >90%)
        """
        total = defects_found_pre_release + defects_found_post_release
        if total == 0:
            return 0
        return (defects_found_pre_release / total) * 100
    
    def defect_density(self, total_defects, size_kloc):
        """
        Defects per 1000 lines of code
        
        Industry average: 3-5 defects per KLOC for good code
        """
        if size_kloc == 0:
            return 0
        return (total_defects / size_kloc) * 1000
    
    def defect_removal_efficiency(self, defects_found_testing, defects_found_before_testing):
        """
        How effective is testing at finding defects?
        DRE = Defects found in Testing / (Defects found in Testing + Defects found after)
        """
        total = defects_found_testing + defects_found_before_testing
        if total == 0:
            return 0
        return (defects_found_testing / total) * 100
    
    def defect_arrival_rate(self, defects_per_week):
        """
        Trend of incoming defects
        Should decrease as release approaches
        """
        import statistics
        avg = statistics.mean(defects_per_week)
        trend = "Increasing" if defects_per_week[-1] > defects_per_week[0] else "Decreasing"
        return {"average": avg, "trend": trend}
    
    def defect_fix_rate(self, fixed_per_week, discovered_per_week):
        """
        Are we fixing faster than discovering?
        """
        if discovered_per_week == 0:
            return "N/A"
        ratio = fixed_per_week / discovered_per_week
        if ratio > 1:
            return "Catching up (Good)"
        elif ratio == 1:
            return "Stable"
        else:
            return "Falling behind (Risk)"
    
    def calculate_all_metrics(self, project_data):
        """
        Comprehensive defect analysis
        """
        metrics = {
            "defect_density": self.defect_density(
                project_data["total_defects"], 
                project_data["code_size_kloc"]
            ),
            "ddp": self.defect_detection_percentage(
                project_data["pre_release_defects"],
                project_data["post_release_defects"]
            ),
            "dre": self.defect_removal_efficiency(
                project_data["testing_defects"],
                project_data["production_defects"]
            ),
            "open_defects": project_data["total_defects"] - project_data["fixed_defects"],
            "defect_age_avg": project_data["total_defect_age_days"] / project_data["fixed_defects"],
            "critical_defects_open": len([d for d in project_data["defects"] if d["severity"] == "Critical" and d["status"] != "Closed"])
        }
        return metrics
```

## **4.4 Testing Metrics and KPIs**

Metrics provide objective measures of test effectiveness, efficiency, and software quality. They enable data-driven decisions about release readiness.

### **4.4.1 Categories of Testing Metrics**

**1. Process Metrics:** Efficiency of testing process (test productivity, effort)
**2. Product Metrics:** Quality of software (defect density, coverage)
**3. Project Metrics:** Progress and status (test execution progress, schedule variance)

```python
class TestingMetrics:
    """
    Comprehensive testing metrics calculations
    """
    
    def __init__(self, project_data):
        self.data = project_data
    
    # ==================== PRODUCT METRICS ====================
    
    def test_coverage_metrics(self):
        """
        How much of the application is tested?
        """
        return {
            "requirement_coverage": {
                "formula": "(Requirements with test cases / Total requirements) * 100",
                "calculation": (self.data["req_with_tests"] / self.data["total_req"]) * 100,
                "target": ">= 95%",
                "interpretation": "Higher is better. Gaps indicate untested functionality."
            },
            "code_coverage": {
                "statement_coverage": self.data["statement_coverage"],
                "branch_coverage": self.data["branch_coverage"],
                "function_coverage": self.data["function_coverage"],
                "path_coverage": self.data["path_coverage"],
                "target": "Statement >= 80%, Branch >= 70%",
                "interpretation": "Percentage of code executed by tests. Higher reduces defect risk."
            },
            "risk_coverage": {
                "formula": "Sum of risk scores of tested requirements / Total risk score",
                "calculation": self.data["tested_risk_score"] / self.data["total_risk_score"],
                "interpretation": "Are high-risk areas thoroughly tested?"
            }
        }
    
    def defect_metrics(self):
        """
        Quality indicators
        """
        return {
            "defect_density": {
                "value": self.data["total_defects"] / self.data["code_size_kloc"],
                "unit": "defects per KLOC",
                "benchmark": "3-5 is industry average, <2 is excellent"
            },
            "defect_severity_index": {
                "formula": "Sum(Severity_weight * Count) / Total_defects",
                "weights": {"Critical": 4, "High": 3, "Medium": 2, "Low": 1},
                "interpretation": "Higher index indicates worse quality"
            },
            "defect_rejection_rate": {
                "formula": "(Invalid defects / Total reported) * 100",
                "target": "< 5%",
                "interpretation": "High rate indicates poor bug reporting or unclear requirements"
            },
            "defect_leakage": {
                "formula": "(Defects found in production / Total defects) * 100",
                "target": "< 5%",
                "interpretation": "Percentage of bugs missed by testing"
            }
        }
    
    # ==================== PROCESS METRICS ====================
    
    def test_execution_metrics(self):
        """
        Efficiency of test execution
        """
        return {
            "test_case_productivity": {
                "formula": "Test cases created / Person-days",
                "value": self.data["test_cases_created"] / self.data["effort_days"],
                "unit": "test cases per day"
            },
            "test_execution_rate": {
                "formula": "Test cases executed / Person-days",
                "value": self.data["test_cases_executed"] / self.data["execution_effort"]
            },
            "test_case_effectiveness": {
                "formula": "(Defects found / Test cases executed) * 100",
                "value": (self.data["defects_found"] / self.data["test_cases_executed"]) * 100,
                "unit": "defects per 100 test cases",
                "interpretation": "Higher indicates better test case quality (finding bugs)"
            },
            "automation_coverage": {
                "formula": "(Automated tests / Total tests) * 100",
                "value": (self.data["automated_tests"] / self.data["total_tests"]) * 100,
                "target": "As high as feasible for regression suite"
            }
        }
    
    def test_effort_metrics(self):
        """
        Cost and efficiency
        """
        total_effort = self.data["test_design_effort"] + self.data["test_execution_effort"]
        
        return {
            "test_effort_percentage": {
                "formula": "(Test effort / Total project effort) * 100",
                "value": (total_effort / self.data["total_project_effort"]) * 100,
                "benchmark": "30-40% for complex projects, 20-30% for simple"
            },
            "cost_per_defect": {
                "formula": "Total testing cost / Defects found",
                "value": self.data["testing_budget"] / self.data["defects_found"],
                "interpretation": "Lower is better, but not if missing critical bugs"
            },
            "automation_roi": {
                "formula": "(Manual effort saved - Automation cost) / Automation cost",
                "calculation": self._calculate_automation_roi(),
                "interpretation": "Positive ROI usually achieved after 3-5 regression cycles"
            }
        }
    
    def _calculate_automation_roi(self):
        manual_effort_per_run = self.data["manual_effort_per_regression"]
        automation_cost = self.data["automation_development_cost"] + self.data["automation_maintenance_cost"]
        runs_per_year = self.data["regression_runs_per_year"]
        
        savings = (manual_effort_per_run * runs_per_year) - automation_cost
        roi = (savings / automation_cost) * 100 if automation_cost > 0 else 0
        return roi
    
    # ==================== PROJECT METRICS ====================
    
    def progress_metrics(self):
        """
        Schedule and progress tracking
        """
        return {
            "test_execution_progress": {
                "formula": "(Executed tests / Planned tests) * 100",
                "value": (self.data["tests_executed"] / self.data["tests_planned"]) * 100,
                "status": "On Track" if self.data["tests_executed"] >= self.data["tests_planned"] * 0.9 else "At Risk"
            },
            "schedule_variance": {
                "formula": "Actual days - Planned days",
                "value": self.data["actual_days"] - self.data["planned_days"],
                "status": "Delayed" if self.data["actual_days"] > self.data["planned_days"] else "On Time"
            },
            "test_estimation_accuracy": {
                "formula": "(Actual effort / Estimated effort) * 100",
                "value": (self.data["actual_effort"] / self.data["estimated_effort"]) * 100,
                "interpretation": "100% is perfect. >100% means underestimated."
            }
        }
    
    def quality_gates(self):
        """
        Pass/Fail criteria for release
        """
        gates = {
            "code_coverage_gate": {
                "requirement": ">= 80%",
                "actual": self.data["code_coverage"],
                "status": "PASS" if self.data["code_coverage"] >= 80 else "FAIL"
            },
            "defect_density_gate": {
                "requirement": "< 5 per KLOC",
                "actual": self.defect_metrics()["defect_density"]["value"],
                "status": "PASS" if self.defect_metrics()["defect_density"]["value"] < 5 else "FAIL"
            },
            "critical_defects_gate": {
                "requirement": "0 open",
                "actual": self.data["open_critical_defects"],
                "status": "PASS" if self.data["open_critical_defects"] == 0 else "FAIL"
            },
            "test_execution_gate": {
                "requirement": ">= 95% executed",
                "actual": self.progress_metrics()["test_execution_progress"]["value"],
                "status": "PASS" if self.progress_metrics()["test_execution_progress"]["value"] >= 95 else "FAIL"
            }
        }
        
        gates["overall_release_readiness"] = "GO" if all(g["status"] == "PASS" for g in gates.values()) else "NO-GO"
        return gates

# Example usage with sample data
sample_project = {
    "total_req": 100,
    "req_with_tests": 98,
    "code_size_kloc": 50,
    "total_defects": 150,
    "statement_coverage": 85,
    "branch_coverage": 75,
    "test_cases_created": 500,
    "effort_days": 25,
    "test_cases_executed": 450,
    "defects_found": 45,
    "automated_tests": 350,
    "total_tests": 500,
    # ... more data
}

metrics_calculator = TestingMetrics(sample_project)
```

## **4.5 Risk-Based Testing Concepts**

Risk-Based Testing (RBT) prioritizes testing efforts based on the likelihood of failure and the impact of that failure. It ensures that limited testing resources are applied where they matter most.

### **4.5.1 Risk Identification and Assessment**

**Risk Formula:**
```
Risk = Probability (Likelihood) × Impact (Severity)
```

**Categories of Risk:**
1. **Product Risk:** Something about the software that might not work (functional defect)
2. **Project Risk:** Something that threatens the testing project (resources, schedule)
3. **Business Risk:** Impact on business operations (reputation, legal compliance)

```python
class RiskBasedTesting:
    """
    Implementing Risk-Based Testing methodology
    """
    
    def risk_assessment_matrix(self):
        """
        5x5 Risk Matrix (Probability x Impact)
        """
        matrix = {
            "Probability": {
                "1_Rare": "< 5% chance",
                "2_Unlikely": "5-20% chance",
                "3_Possible": "20-50% chance",
                "4_Likely": "50-80% chance",
                "5_Almost_Certain": "> 80% chance"
            },
            "Impact": {
                "1_Negligible": "Minimal impact, workaround easy",
                "2_Minor": "Low impact, some inconvenience",
                "3_Moderate": "Significant impact, workaround exists",
                "4_Major": "High impact, no workaround, financial loss",
                "5_Catastrophic": "System down, data loss, legal issues"
            }
        }
        
        # Risk Score = Probability * Impact
        # Risk Levels:
        risk_levels = {
            "1-4": "Low (Green) - Minimal testing, sample only",
            "5-9": "Medium (Yellow) - Standard testing effort",
            "10-19": "High (Orange) - Extensive testing required",
            "20-25": "Critical (Red) - Exhaustive testing, management attention"
        }
        
        return matrix, risk_levels
    
    def identify_product_risks(self, application_context):
        """
        Systematic risk identification
        """
        risk_categories = {
            "functional_risks": [
                "Complex calculations (financial, scientific)",
                "Multi-step workflows (checkout, registration)",
                "Integration points with external systems",
                "Authentication and authorization",
                "Data migration and conversion"
            ],
            "technical_risks": [
                "New technology (unproven in organization)",
                "High complexity modules (cyclomatic complexity > 20)",
                "Legacy code with no unit tests",
                "Concurrent processing/threading",
                "Memory-intensive operations"
            ],
            "business_risks": [
                "Revenue-generating features",
                "Regulatory compliance features (GDPR, SOX)",
                "Customer-facing vs. internal tools",
                "High-usage features (80/20 rule)",
                "Competitive differentiators"
            ],
            "external_risks": [
                "Third-party API dependencies",
                "Browser/device compatibility",
                "Network latency sensitivity",
                "Security vulnerabilities (OWASP Top 10)",
                "Performance under load"
            }
        }
        
        # Example risk identification for e-commerce
        if application_context == "e-commerce":
            identified_risks = [
                {
                    "id": "R001",
                    "description": "Payment processing failure",
                    "probability": 3,  # Possible (complex integration)
                    "impact": 5,       # Catastrophic (revenue loss)
                    "risk_score": 15,  # High
                    "mitigation": "Extensive integration testing, fallback mechanisms"
                },
                {
                    "id": "R002",
                    "description": "Search returns wrong results",
                    "probability": 4,  # Likely (complex query logic)
                    "impact": 3,       # Moderate (annoyance, lost sales)
                    "risk_score": 12,  # High
                    "mitigation": "Comprehensive search test suite, relevance testing"
                },
                {
                    "id": "R003",
                    "description": "About us page typo",
                    "probability": 2,  # Unlikely
                    "impact": 1,       # Negligible
                    "risk_score": 2,   # Low
                    "mitigation": "Minimal testing, visual check only"
                }
            ]
            return identified_risks
        
        return risk_categories
    
    def prioritize_test_cases_by_risk(self, test_cases, risks):
        """
        Map test cases to risks and prioritize
        """
        prioritized_tests = []
        
        for test in test_cases:
            # Calculate risk weight of test case
            risk_weight = 0
            related_risks = []
            
            for risk in risks:
                if self._test_covers_risk(test, risk):
                    risk_weight += risk["risk_score"]
                    related_risks.append(risk["id"])
            
            test["risk_weight"] = risk_weight
            test["related_risks"] = related_risks
            test["priority"] = self._calculate_priority_from_risk(risk_weight)
            prioritized_tests.append(test)
        
        # Sort by risk weight descending
        return sorted(prioritized_tests, key=lambda x: x["risk_weight"], reverse=True)
    
    def _test_covers_risk(self, test_case, risk):
        """
        Determine if a test case mitigates a specific risk
        """
        # Logic to map test to risk based on component, function, etc.
        return risk["component"] in test_case.get("components", [])
    
    def _calculate_priority_from_risk(self, risk_weight):
        if risk_weight >= 15:
            return "Critical"
        elif risk_weight >= 10:
            return "High"
        elif risk_weight >= 5:
            return "Medium"
        else:
            return "Low"

    def risk_based_test_strategy(self):
        """
        Allocating testing effort based on risk
        """
        strategy = {
            "critical_risk_20_25": {
                "test_intensity": "Exhaustive",
                "techniques": ["Formal reviews", "Unit testing", "Integration testing", 
                              "System testing", "Regression testing", "Exploratory testing"],
                "automation": "100% automated regression",
                "resources": "Senior testers only",
                "documentation": "Detailed evidence required"
            },
            "high_risk_10_19": {
                "test_intensity": "Thorough",
                "techniques": ["Unit testing", "Integration testing", "System testing"],
                "automation": "Automated critical path",
                "resources": "Experienced testers",
                "documentation": "Standard reporting"
            },
            "medium_risk_5_9": {
                "test_intensity": "Standard",
                "techniques": ["System testing", "Sample regression"],
                "automation": "Automated smoke tests",
                "resources": "Standard team",
                "documentation": "Basic reporting"
            },
            "low_risk_1_4": {
                "test_intensity": "Minimal",
                "techniques": ["High-level validation", "Sample testing"],
                "automation": "Optional",
                "resources": "Junior testers or automated",
                "documentation": "Minimal"
            }
        }
        return strategy
```
