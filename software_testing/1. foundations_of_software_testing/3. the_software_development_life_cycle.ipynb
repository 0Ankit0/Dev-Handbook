## **3.1 Introduction to the Software Development Life Cycle**

### **What is the SDLC?**

The Software Development Life Cycle (SDLC) is a structured process that defines the steps involved in developing software from initial conception to final deployment and maintenance. Think of it as the "blueprint" for building software—just as architects follow specific phases to construct a building (planning, foundation, framing, finishing), software teams follow the SDLC to create applications.

**Formal Definition:**
> "The Software Development Life Cycle (SDLC) is a process used by the software industry to design, develop, and test high-quality software. It aims to produce software that meets or exceeds customer expectations, reaches completion within times and cost estimates." — IEEE

### **Why SDLC Matters for Testers**

As a tester, understanding the SDLC is crucial because:

1. **Testing is not a phase—it's an activity** that occurs throughout the lifecycle
2. **Your role changes** depending on which phase the project is in
3. **The testing approach** must adapt to the chosen SDLC model (Waterfall vs. Agile require different strategies)
4. **Early involvement** (Shift-Left) is only possible if you understand when and where to insert testing activities
5. **Deliverables and expectations** vary by phase and model

### **Generic SDLC Phases**

Regardless of the specific model (Waterfall, Agile, etc.), all SDLCs generally include these phases:

```
┌─────────────────────────────────────────────────────────────┐
│                 SOFTWARE DEVELOPMENT LIFE CYCLE              │
├─────────────────────────────────────────────────────────────┤
│                                                              │
│  1. Planning/Requirements  →  "What should we build?"       │
│            │                                                 │
│            ▼                                                 │
│  2. Analysis/Design        →  "How should we build it?"     │
│            │                                                 │
│            ▼                                                 │
│  3. Implementation         →  "Let's build it!"             │
│            │                                                 │
│            ▼                                                 │
│  4. Testing                →  "Did we build it right?"      │
│            │                                                 │
│            ▼                                                 │
│  5. Deployment             →  "Release it!"                 │
│            │                                                 │
│            ▼                                                 │
│  6. Maintenance            →  "Keep it running!"            │
│                                                              │
└─────────────────────────────────────────────────────────────┘
```

Let's examine each phase in detail from a testing perspective:

#### **Phase 1: Planning and Requirements Analysis**

**Activities:**
- Gathering business requirements
- Feasibility studies
- Resource allocation
- Project scope definition

**Testing Activities (Shift-Left):**
- Review requirements for testability
- Identify testable vs. non-testable requirements
- Begin test planning
- Risk assessment

**Key Deliverables:**
- Software Requirement Specification (SRS)
- Test feasibility report
- Initial test strategy

```python
# Example: Requirements Review Checklist for Testers

class RequirementsReview:
    """
    Tester's checklist for reviewing requirements
    """
    
    def __init__(self, requirements_doc):
        self.requirements = requirements_doc
        self.issues = []
    
    def check_clarity(self):
        """
        Verify requirements are clear and unambiguous
        """
        checks = {
            "clear_language": "Requirements use clear, concise language",
            "no_jargon": "Technical terms are defined",
            "measurable": "Requirements have measurable criteria",
            "consistent": "No contradictions between requirements",
            "complete": "All necessary information is included"
        }
        
        # Example of finding an ambiguous requirement
        ambiguous_req = "The system should be fast"
        # Tester flags this: What does "fast" mean? 
        # Should be: "The system shall respond to user input within 2 seconds"
        
        return checks
    
    def check_testability(self):
        """
        Verify requirements can be tested
        """
        testability_criteria = [
            "Specific: The requirement states exact behavior",
            "Measurable: Success can be quantified", 
            "Achievable: Can be implemented and tested",
            "Relevant: Aligns with business needs",
            "Time-bound: Has deadlines or version targets"
        ]
        
        # Example testable requirement:
        good_example = """
        REQ-001: The login system shall lock the user account 
        after 5 consecutive failed attempts within a 15-minute window.
        
        Testable: Yes
        - Can test 4 attempts (should not lock)
        - Can test 5 attempts (should lock)
        - Can test timing (15-minute window)
        """
        
        return testability_criteria
    
    def identify_test_scenarios(self):
        """
        Extract initial test scenarios from requirements
        """
        scenarios = []
        
        # From requirement: "User shall be able to reset password"
        scenarios.extend([
            "Test: Valid email password reset",
            "Test: Invalid email handling", 
            "Test: Expired reset link",
            "Test: Reusing old reset link",
            "Test: Reset email delivery"
        ])
        
        return scenarios
```

#### **Phase 4: Testing (Verification & Validation)**

**Activities:**
- Integration testing
- System testing
- User Acceptance Testing (UAT)
- Performance testing
- Security testing

**Testing Activities:**
- Execute test cases
- Report and track defects
- Regression testing
- Test reporting
- Release recommendation

**Key Deliverables:**
- Test execution reports
- Defect reports
- Test summary report
- Release decision (Go/No-Go)

```python
# Example: Test Execution and Reporting

class TestExecution:
    def __init__(self):
        self.test_results = []
        self.defects = []
    
    def execute_test_cycle(self, test_cases, build_version):
        """
        Execute a full test cycle
        """
        cycle_results = {
            "build_version": build_version,
            "start_time": "2026-01-15T09:00:00Z",
            "test_cases": [],
            "summary": {}
        }
        
        for tc in test_cases:
            # Simulate test execution
            result = self.run_test(tc)
            
            cycle_results["test_cases"].append({
                "id": tc["id"],
                "status": result["status"],  # Pass/Fail/Blocked
                "duration": result["duration"],
                "defect_id": result.get("defect_id"),
                "evidence": result.get("screenshot_path")
            })
        
        # Calculate metrics
        total = len(test_cases)
        passed = sum(1 for r in cycle_results["test_cases"] if r["status"] == "Pass")
        failed = sum(1 for r in cycle_results["test_cases"] if r["status"] == "Fail")
        
        cycle_results["summary"] = {
            "total": total,
            "passed": passed,
            "failed": failed,
            "pass_rate": (passed/total)*100,
            "recommendation": "Proceed" if failed == 0 else "Fix defects before release"
        }
        
        return cycle_results
    
    def generate_release_recommendation(self, test_results):
        """
        Provide Go/No-Go decision based on test results
        """
        criteria = {
            "critical_defects": 0,
            "major_defects": "<= 2",
            "pass_rate": ">= 95%",
            "performance_regression": "None"
        }
        
        actual = {
            "critical_defects": test_results.get("critical_count", 0),
            "major_defects": test_results.get("major_count", 0),
            "pass_rate": test_results.get("pass_rate", 0)
        }
        
        if actual["critical_defects"] == 0 and actual["pass_rate"] >= 95:
            return {
                "decision": "GO",
                "risk_level": "Low",
                "conditions": "Standard release process"
            }
        else:
            return {
                "decision": "NO-GO",
                "risk_level": "High",
                "conditions": "Fix critical defects and retest"
            }
```

## **3.2 SDLC Models: Traditional Approaches**

Different projects require different approaches. Let's explore the major SDLC models and how testing fits into each.

### **3.2.1 The Waterfall Model**

**Overview:**
The Waterfall model is a linear, sequential approach where each phase must be completed before the next begins. It's like a waterfall—water flows down step by step and doesn't go back up.

```
Waterfall Model:

Requirements → Design → Implementation → Testing → Deployment → Maintenance
      │            │           │            │           │            │
      ▼            ▼           ▼            ▼           ▼            ▼
   [Frozen]    [Frozen]    [Coding]    [Testing]   [Release]   [Support]
   
   Phase 1      Phase 2     Phase 3     Phase 4     Phase 5     Phase 6
   
Characteristics:
- Each phase has defined start and end
- Documentation-heavy
- Changes are difficult and expensive
- Testing happens late (Phase 4)
```

**When to Use:**
- Requirements are well-understood and unlikely to change
- Project is simple and short
- Industry with strict regulatory requirements (medical devices, aerospace)
- Fixed-price contracts with clear scope

**Testing in Waterfall:**

```python
class WaterfallTesting:
    """
    Testing approach in Waterfall model
    """
    
    def phase_1_requirements(self):
        """
        Testing activities in Requirements phase
        """
        return {
            "activities": [
                "Review Software Requirements Specification (SRS)",
                "Identify testable requirements",
                "Create Requirements Traceability Matrix (RTM)",
                "Identify test data requirements",
                "Estimate testing effort"
            ],
            "deliverables": [
                "Requirements Review Report",
                "Initial Test Plan",
                "RTM (initial version)"
            ]
        }
    
    def phase_2_design(self):
        """
        Testing activities in Design phase  
        """
        return {
            "activities": [
                "Review High-Level Design (HLD)",
                "Review Low-Level Design (LLD)",
                "Identify integration test scenarios",
                "Design test architecture",
                "Plan test environments"
            ],
            "deliverables": [
                "Design Review Report",
                "Test Architecture Document",
                "Test Environment Plan"
            ]
        }
    
    def phase_3_implementation(self):
        """
        Testing activities in Implementation phase
        """
        return {
            "activities": [
                "Develop detailed test cases",
                "Prepare test data",
                "Set up test environments",
                "Review unit test plans",
                "Create test scripts"
            ],
            "deliverables": [
                "Test Cases (reviewed and approved)",
                "Test Data",
                "Test Environment (ready)",
                "Test Scripts"
            ]
        }
    
    def phase_4_testing(self):
        """
        The main testing phase
        """
        return {
            "levels": [
                {
                    "level": "Unit Testing",
                    "owner": "Developers",
                    "focus": "Individual components"
                },
                {
                    "level": "Integration Testing", 
                    "owner": "Testers/Developers",
                    "focus": "Component interfaces"
                },
                {
                    "level": "System Testing",
                    "owner": "QA Team",
                    "focus": "End-to-end functionality"
                },
                {
                    "level": "User Acceptance Testing",
                    "owner": "Business Users",
                    "focus": "Business requirements"
                }
            ],
            "entry_criteria": [
                "Code development complete",
                "Unit testing complete",
                "Test cases reviewed and approved",
                "Test environment ready"
            ],
            "exit_criteria": [
                "All test cases executed",
                "Critical and High defects fixed",
                "Test summary report approved",
                "Acceptance criteria met"
            ]
        }
    
    def pros_cons(self):
        return {
            "pros": [
                "Simple and easy to understand",
                "Clear milestones and deliverables",
                "Well-documented process",
                "Easy to manage due to rigidity"
            ],
            "cons": [
                "Testing happens late - expensive to fix bugs",
                "No working software until late in the cycle",
                "Difficult to accommodate changing requirements",
                "High risk and uncertainty"
            ]
        }
```

**Waterfall Testing Challenges:**

1. **Late Bug Detection:** If requirements were misunderstood in Phase 1, you don't find out until Phase 4—expensive!
2. **No Feedback Loop:** Users don't see anything until the end
3. **Tester's Role Limited:** Testers are siloed until testing phase

### **3.2.2 The V-Model (Verification and Validation)**

**Overview:**
The V-Model extends the Waterfall model by emphasizing the relationship between development phases and testing phases. It forms a "V" shape where verification (left side) corresponds to validation (right side).

```
The V-Model:

          Verification                Validation
          (Building it right)         (Right product)
          
Requirements    │    Acceptance Testing
   Analysis     │         ▲
      │         │         │ (Validate requirements)
      ▼         │         │
 System Design  │    System Testing
      │         │         ▲
      ▼         │         │ (Validate design)
Architecture    │         │
      │         │         │
      ▼         │         │
 Module Design  │    Integration Testing
      │         │         ▲
      ▼         │         │ (Validate modules)
   Coding       │         │
      │         │         │
      └─────────┴─────────┘
                │
        Unit Testing
                ▲
                │ (Validate code)
                
Left side: Development activities (Decomposition)
Right side: Testing activities (Integration/Validation)
```

**The Correspondence:**

| Development Phase | Testing Phase | Focus |
|-------------------|---------------|-------|
| Requirements Analysis | Acceptance Testing | Validate business needs |
| System Design | System Testing | Validate system behavior |
| Architecture Design | Integration Testing | Validate component integration |
| Module Design | Unit Testing | Validate individual modules |

**Advantage over Waterfall:** Testing is planned early (test cases designed during corresponding dev phase) even though execution happens later.

```python
class VModelTesting:
    """
    Testing activities in V-Model aligned with development phases
    """
    
    def requirements_phase(self):
        """
        Left side: Requirements
        Right side: Acceptance Test Planning
        """
        return {
            "development_activity": "Gather and document requirements",
            "testing_activity": "Plan Acceptance Tests",
            "test_deliverables": [
                "Acceptance Test Plan",
                "Acceptance Criteria Definition",
                "UAT Test Cases (high-level)"
            ],
            "focus": "Does the system meet business needs?"
        }
    
    def system_design_phase(self):
        """
        Left side: System Design
        Right side: System Test Planning
        """
        return {
            "development_activity": "Design system architecture",
            "testing_activity": "Design System Tests",
            "test_deliverables": [
                "System Test Plan",
                "System Test Cases",
                "Test Environment Specification"
            ],
            "focus": "Does the system meet functional & non-functional requirements?"
        }
    
    def architecture_design_phase(self):
        """
        Left side: Architecture/Component Design
        Right side: Integration Test Planning
        """
        return {
            "development_activity": "Design components and interfaces",
            "testing_activity": "Design Integration Tests",
            "test_deliverables": [
                "Integration Test Plan",
                "Integration Test Cases",
                "Stub and Driver Specifications"
            ],
            "focus": "Do components work together correctly?"
        }
    
    def module_design_phase(self):
        """
        Left side: Module/Detailed Design
        Right side: Unit Test Planning
        """
        return {
            "development_activity": "Design individual modules",
            "testing_activity": "Design Unit Tests",
            "test_deliverables": [
                "Unit Test Plans",
                "Unit Test Cases",
                "Code Review Checklists"
            ],
            "focus": "Does each module work correctly in isolation?"
        }
    
    def coding_phase(self):
        """
        Bottom of the V: Implementation
        """
        return {
            "activity": "Write code based on detailed designs",
            "parallel_testing": "Begin writing test scripts based on test designs",
            "deliverables": [
                "Source Code",
                "Unit Test Scripts",
                "Technical Documentation"
            ]
        }
    
    def testing_execution_side(self):
        """
        Right side of V: Testing Execution (upward)
        """
        phases = [
            {
                "phase": "Unit Testing",
                "responsible": "Developers",
                "basis": "Module Design",
                "technique": "White-box testing"
            },
            {
                "phase": "Integration Testing",
                "responsible": "Developers/Testers",
                "basis": "Architecture Design",
                "technique": "Top-down, Bottom-up, Big-bang"
            },
            {
                "phase": "System Testing",
                "responsible": "Testers",
                "basis": "System Design",
                "technique": "Black-box testing"
            },
            {
                "phase": "Acceptance Testing",
                "responsible": "Users/Business Analysts",
                "basis": "Requirements",
                "technique": "UAT, Alpha, Beta testing"
            }
        ]
        return phases
```

### **3.2.3 Iterative and Incremental Models**

**Overview:**
Instead of delivering everything at once, the software is developed and delivered in increments (pieces). Each iteration goes through mini-SDLC phases.

```
Iterative Model:

Iteration 1:          Iteration 2:         Iteration 3:
┌──────────┐         ┌──────────┐         ┌──────────┐
│ Req      │         │ Req      │         │ Req      │
│ Design   │         │ Design   │         │ Design   │
│ Code     │         │ Code     │         │ Code     │
│ Test     │         │ Test     │         │ Test     │
└──────────┘         └──────────┘         └──────────┘
     │                    │                    │
     ▼                    ▼                    ▼
[Core Features]      [More Features]      [Complete Product]
  (Basic)              (Enhanced)           (Polished)
  
Risk is spread across iterations. Early iterations identify issues before too much investment.
```

**Testing in Iterative Model:**

- Each iteration has its own complete testing cycle
- Regression testing crucial (ensure new iterations don't break old ones)
- Continuous integration of components

```python
class IterativeTesting:
    def iteration_plan(self, iteration_number, features):
        """
        Testing for a specific iteration
        """
        plan = {
            "iteration": iteration_number,
            "scope": features,
            "testing_activities": [
                "Requirements review for this iteration's features",
                "Test planning for new features",
                "Test case development",
                "Functional testing of new features",
                "Regression testing of previous features",
                "Integration testing",
                "User demo/feedback"
            ],
            "entry_criteria": [
                "Previous iteration features stable",
                "New iteration requirements approved"
            ],
            "exit_criteria": [
                "New features tested and working",
                "No critical defects",
                "Regression tests pass",
                "Stakeholder approval"
            ]
        }
        return plan
    
    def regression_strategy(self):
        """
        As system grows, regression testing becomes critical
        """
        strategy = {
            "iteration_1": "Test all new features (100% new)",
            "iteration_2": "Test new features + regression of iteration 1",
            "iteration_3": "Test new features + regression of iteration 1-2",
            "automation_approach": "Automate previous iteration tests to handle growing regression suite"
        }
        return strategy
```

## **3.3 Modern SDLC Models: Agile and DevOps**

### **3.3.1 Agile Methodology**

**Overview:**
Agile is an iterative approach that emphasizes:
- **Individuals and interactions** over processes and tools
- **Working software** over comprehensive documentation  
- **Customer collaboration** over contract negotiation
- **Responding to change** over following a plan

**Popular Agile Frameworks:**
- **Scrum:** Time-boxed sprints (2-4 weeks), defined roles (Scrum Master, Product Owner, Team)
- **Kanban:** Continuous flow, visual board, work-in-progress limits
- **Extreme Programming (XP):** Engineering practices like pair programming, TDD

**Testing in Agile:**

```python
class AgileTesting:
    """
    Testing in Agile/Scrum environment
    """
    
    def sprint_cycle(self, sprint_duration="2 weeks"):
        """
        Testing activities throughout a sprint
        """
        timeline = {
            "Day 1 (Sprint Planning)": {
                "testing_tasks": [
                    "Review user stories for testability",
                    "Point stories (estimate effort)",
                    "Identify test data needs",
                    "Plan automation for the sprint"
                ]
            },
            "Day 2-8 (Development)": {
                "testing_tasks": [
                    "Write acceptance criteria (Given-When-Then)",
                    "Create test cases in parallel with development",
                    "Set up test data",
                    "Begin exploratory testing",
                    "Pair with developers",
                    "Automate tests for completed features"
                ]
            },
            "Day 9-10 (Testing/Validation)": {
                "testing_tasks": [
                    "Execute test cases",
                    "Perform exploratory testing",
                    "Bug reporting and verification",
                    "Regression testing",
                    "Demo preparation"
                ]
            },
            "Sprint Review (Day 10)": {
                "testing_tasks": [
                    "Demo working features",
                    "Report quality metrics",
                    "Gather feedback for next sprint"
                ]
            },
            "Sprint Retrospective": {
                "testing_tasks": [
                    "Discuss testing process improvements",
                    "Identify automation gaps",
                    "Plan testing improvements"
                ]
            }
        }
        return timeline
    
    def agile_testing_quadrants(self):
        """
        Brian Marick's Agile Testing Quadrants
        """
        quadrants = {
            "Q1_Technology_Facing_Support_Development": {
                "focus": "Unit tests, Component tests",
                "automation": "Automated",
                "purpose": "Guide development",
                "examples": ["JUnit tests", "TDD", "API tests"]
            },
            "Q2_Business_Facing_Support_Development": {
                "focus": "Functional tests, Examples, Story tests",
                "automation": "Automated & Manual",
                "purpose": "Guide development",
                "examples": ["BDD scenarios", "Prototype tests", "Pair testing"]
            },
            "Q3_Business_Facing_Critique_Product": {
                "focus": "Exploratory testing, Scenarios, Usability",
                "automation": "Manual",
                "purpose": "Evaluate product",
                "examples": ["User acceptance", "Alpha/Beta", "Exploratory sessions"]
            },
            "Q4_Technology_Facing_Critique_Product": {
                "focus": "Performance, Security, Stability",
                "automation": "Tools",
                "purpose": "Evaluate product",
                "examples": ["Load testing", "Security scans", "Memory leaks"]
            }
        }
        return quadrants
    
    def definition_of_done(self):
        """
        Testing criteria for a story to be considered "Done"
        """
        dod = [
            "Code is written and peer-reviewed",
            "Unit tests written and passing (>80% coverage)",
            "Integration tests passing",
            "Acceptance criteria met",
            "Tested on all supported browsers/devices",
            "No critical or high defects",
            "Documentation updated",
            "Product Owner acceptance",
            "Automated tests added to regression suite"
        ]
        return dod

    def continuous_testing_in_agile(self):
        """
        Testing happens continuously, not at the end
        """
        pipeline = {
            "Pre-commit": {
                "tests": ["Unit tests", "Static analysis"],
                "duration": "< 2 minutes",
                "responsibility": "Developer"
            },
            "Commit": {
                "tests": ["Integration tests", "API tests"],
                "duration": "5-10 minutes",
                "responsibility": "CI Server"
            },
            "Build": {
                "tests": ["Smoke tests", "Security scan"],
                "duration": "10-15 minutes",
                "responsibility": "CI/CD Pipeline"
            },
            "Staging": {
                "tests": ["E2E tests", "Performance baseline", "UAT"],
                "duration": "30-60 minutes",
                "responsibility": "QA Team"
            },
            "Production": {
                "tests": ["Smoke tests", "Monitoring", "A/B tests"],
                "duration": "Continuous",
                "responsibility": "DevOps/QA"
            }
        }
        return pipeline
```

**Agile Testing Principles (from Lisa Crispin and Janet Gregory):**

1. **Provide continuous feedback** (not just at the end)
2. **Deliver value to the customer** (test what matters)
3. **Enable face-to-face communication** (collaboration over documentation)
4. **Have courage** (speak up about quality issues)
5. **Keep it simple** (just enough testing)
6. **Practice continuous improvement** (retrospectives)
7. **Respond to change** (adapt test strategy)
8. **Self-organize** (team decides how to test)
9. **Focus on people** (customer satisfaction)
10. **Enjoy!** (sustainable pace)

## **3.4 The Software Testing Life Cycle (STLC)**

While SDLC covers the entire software development, STLC (Software Testing Life Cycle) focuses specifically on the testing phases. STLC can occur within any SDLC model.

### **STLC Phases:**

```
Software Testing Life Cycle:

1. Requirement Analysis ───────► Input: SRS, BRS
        │
        ▼
2. Test Planning ──────────────► Output: Test Plan, Strategy
        │
        ▼
3. Test Case Development ──────► Output: Test Cases, Scripts
        │
        ▼
4. Test Environment Setup ─────► Output: Test Bed, Data
        │
        ▼
5. Test Execution ─────────────► Output: Test Results, Defects
        │
        ▼
6. Test Cycle Closure ─────────► Output: Test Summary Report
```

### **Detailed STLC Activities:**

```python
class STLC:
    """
    Software Testing Life Cycle implementation
    """
    
    def phase_1_requirement_analysis(self):
        """
        Understand what to test
        """
        activities = {
            "inputs": [
                "Software Requirements Specification (SRS)",
                "Business Requirements Document (BRD)",
                "Use Cases",
                "User Stories"
            ],
            "activities": [
                "Review requirements documents",
                "Identify scope of testing",
                "Identify testable vs. non-testable requirements",
                "Create Requirements Traceability Matrix (RTM)",
                "Identify types of testing needed"
            ],
            "deliverables": [
                "RTM (Requirements Traceability Matrix)",
                "Automation feasibility report",
                "List of questions/clarifications"
            ],
            "exit_criteria": [
                "All requirements reviewed",
                "RTM created",
                "Scope clearly defined"
            ]
        }
        return activities
    
    def phase_2_test_planning(self):
        """
        Plan how to test
        """
        activities = {
            "inputs": [
                "Requirements documents",
                "RTM",
                "Project plan"
            ],
            "activities": [
                "Define test strategy",
                "Estimate effort and resources",
                "Identify test tools",
                "Define test environment requirements",
                "Define entry and exit criteria",
                "Identify risks and mitigation",
                "Create test schedule"
            ],
            "deliverables": [
                "Test Plan Document",
                "Test Strategy Document",
                "Test Estimation",
                "Resource allocation plan"
            ],
            "exit_criteria": [
                "Test Plan approved by stakeholders",
                "Resources allocated",
                "Schedule baselined"
            ]
        }
        return activities
    
    def phase_3_test_case_development(self):
        """
        Create test cases
        """
        activities = {
            "inputs": [
                "Requirements",
                "RTM",
                "Test Plan"
            ],
            "activities": [
                "Create test scenarios",
                "Write detailed test cases",
                "Review test cases",
                "Create test data",
                "Prepare test scripts (automation)",
                "Create test traceability"
            ],
            "deliverables": [
                "Test Cases (documented)",
                "Test Scripts (automation)",
                "Test Data",
                "RTM (updated with test cases)"
            ],
            "exit_criteria": [
                "Test cases reviewed and approved",
                "Test data prepared",
                "Automation scripts ready"
            ]
        }
        return activities
    
    def phase_4_test_environment_setup(self):
        """
        Prepare where to test
        """
        activities = {
            "inputs": [
                "Environment requirements",
                "Architecture documents"
            ],
            "activities": [
                "Set up hardware/software",
                "Install test tools",
                "Configure network",
                "Create test database",
                "Populate test data",
                "Smoke test environment"
            ],
            "deliverables": [
                "Test Environment (ready)",
                "Environment configuration document",
                "Smoke test results"
            ],
            "exit_criteria": [
                "Environment ready and stable",
                "Smoke tests pass",
                "Access granted to team"
            ]
        }
        return activities
    
    def phase_5_test_execution(self):
        """
        Execute tests
        """
        activities = {
            "inputs": [
                "Test cases",
                "Test environment",
                "Build to test"
            ],
            "activities": [
                "Execute test cases",
                "Record results (Pass/Fail/Blocked)",
                "Report defects",
                "Retest fixed defects",
                "Regression testing",
                "Update RTM with execution status"
            ],
            "deliverables": [
                "Test Execution Report",
                "Defect Reports",
                "Updated RTM",
                "Test Logs"
            ],
            "exit_criteria": [
                "All test cases executed (or deferred with reason)",
                "Defects reported and tracked",
                "Exit criteria met (or approved exception)"
            ]
        }
        return activities
    
    def phase_6_test_closure(self):
        """
        Wrap up testing
        """
        activities = {
            "inputs": [
                "Test execution reports",
                "Defect reports",
                "RTM"
            ],
            "activities": [
                "Evaluate exit criteria",
                "Prepare test summary report",
                "Document lessons learned",
                "Archive test assets",
                "Handover to maintenance team",
                "Celebrate success!"
            ],
            "deliverables": [
                "Test Summary Report",
                "Lessons Learned Document",
                "Release Recommendation (Go/No-Go)",
                "Archived test artifacts"
            ],
            "exit_criteria": [
                "Test Summary Report approved",
                "Knowledge transfer complete",
                "Sign-off obtained"
            ]
        }
        return activities
```

## **3.5 Entry and Exit Criteria**

Entry and exit criteria define when a phase can start (entry) and when it can be considered complete (exit).

```python
class EntryExitCriteria:
    """
    Entry and Exit criteria for testing phases
    """
    
    def test_planning_entry_criteria(self):
        return [
            "Requirements documents approved and available",
            "Project plan and schedule defined",
            "Roles and responsibilities defined",
            "Test tools identified and procured"
        ]
    
    def test_planning_exit_criteria(self):
        return [
            "Test Plan document signed off",
            "Test Strategy approved",
            "Resource allocation confirmed",
            "Test schedule baselined"
        ]
    
    def test_execution_entry_criteria(self):
        return [
            "Test cases reviewed and approved",
            "Test environment ready and smoke tested",
            "Build deployed in test environment",
            "Test data prepared",
            "Entry criteria defined in Test Plan are met"
        ]
    
    def test_execution_exit_criteria(self):
        return [
            "All test cases executed (or explicitly deferred)",
            "Critical and High severity defects closed",
            "Test coverage meets defined threshold (e.g., 90%)",
            "No open blocking issues",
            "Defect density below threshold (e.g., < 5 per 1000 LOC)",
            "Performance criteria met",
            "Stakeholder sign-off obtained"
        ]
    
    def suspension_criteria(self):
        """
        When to pause testing
        """
        return [
            "Build is unstable (smoke tests fail)",
            "Critical blocking defect found",
            "Test environment unavailable",
            "Major requirement change (scope change)",
            "Resources unavailable (illness, reassignment)"
        ]
    
    def resumption_criteria(self):
        """
        When to resume testing
        """
        return [
            "New stable build available",
            "Blocking defect fixed and verified",
            "Environment restored",
            "Change request approved and scope updated",
            "Resources available"
        ]
```

## **Chapter Summary**

### **Key Takeaways from Chapter 3:**

**SDLC Fundamentals:**
- The SDLC provides a structured framework for software development from conception to retirement
- Testing is not a single phase but an activity integrated throughout the lifecycle
- Understanding SDLC helps testers know when and how to engage

**Traditional Models:**
- **Waterfall:** Linear, sequential, testing at the end. Good for fixed-scope projects but risky for bug detection.
- **V-Model:** Corresponds development phases with testing phases. Better planning but still late testing.
- **Iterative:** Delivers in increments, reducing risk through early feedback.

**Modern Approaches:**
- **Agile:** Iterative, collaborative, continuous testing within sprints. Testers are integrated team members.
- **DevOps:** Extends Agile with continuous integration/delivery, automated testing pipelines, and shift-left practices.

**STLC (Software Testing Life Cycle):**
1. **Requirement Analysis:** Understand what to test
2. **Test Planning:** Plan how to test  
3. **Test Case Development:** Create test cases and scripts
4. **Test Environment Setup:** Prepare test bed
5. **Test Execution:** Run tests and report defects
6. **Test Closure:** Summarize and archive

**Entry/Exit Criteria:**
- Define clear conditions for starting and completing each phase
- Suspension/Resumption criteria handle interruptions
- Ensures quality gates are maintained

**Industry Standards:**
- IEEE 1074: Standard for Developing Software Life Cycle Processes
- ISO/IEC 12207: Software life cycle processes
- ISO/IEC 25010: Software product quality
