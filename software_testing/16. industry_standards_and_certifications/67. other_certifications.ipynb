# Chapter 67: Other Certifications

---

## 67.1 Introduction

While ISTQB is the most widely recognized certification for software testing professionals, it is by no means the only one. A variety of other certifications exist, offered by different organizations, each focusing on specific aspects of testing, tools, domains, or methodologies. Some are general and vendor-neutral, others are tied to specific tools or platforms, and yet others address specialized areas like security or agile testing. Choosing the right certification(s) can enhance your career, demonstrate expertise, and open doors to new opportunities.

### 67.1.1 Why Pursue Non-ISTQB Certifications?

- **Specialization:** ISTQB provides a broad foundation; other certifications allow you to dive deep into specific areas (e.g., test automation, performance testing, security).
- **Vendor Recognition:** If your organization uses a particular tool (e.g., Selenium, Tosca, LoadRunner), vendor certifications validate your proficiency.
- **Domain Expertise:** Certifications in agile, DevOps, or security complement testing skills and make you more versatile.
- **Career Differentiation:** A mix of certifications can set you apart from candidates with only ISTQB.

---

## 67.2 General Testing Certifications (Non-ISTQB)

Several organizations offer general testing certifications that are not part of the ISTQB scheme. Some are regionally focused, while others are international.

### 67.2.1 CSTE (Certified Software Tester)

Offered by **QAI Global Institute**, CSTE is a well-known certification for experienced testing professionals. It focuses on the principles, practices, and techniques of software testing, with an emphasis on management and process improvement.

- **Prerequisites:** Typically requires 2 years of experience in testing (varies by educational background).
- **Exam Format:** Two parts – multiple choice and essay questions. The exam covers topics like testing concepts, test management, test design, defect tracking, and metrics.
- **Value:** Recognized globally, especially in North America and parts of Asia. It is often compared to ISTQB Advanced Level.
- **Cost:** Approximately $400–$500 (varies by region).

### 67.2.2 CAST (Certified Associate in Software Testing)

Also from QAI, CAST is an entry-level certification for those new to testing or with less than two years of experience.

- **Prerequisites:** None.
- **Exam Format:** Multiple choice questions covering fundamentals similar to ISTQB Foundation but with a slightly different emphasis.
- **Value:** Good starting point for those who want a certification but aren't yet eligible for ISTQB or CSTE.
- **Cost:** Around $300–$400.

### 67.2.3 ISEB/BCS Certifications

The **British Computer Society (BCS)** offers a range of testing certifications, including:

- **BCS Foundation Certificate in Software Testing** (equivalent to ISTQB Foundation)
- **BCS Intermediate Certificate in Software Testing**
- **BCS Practitioner Certificates** (e.g., Test Management, Test Analysis, Performance Testing)

These are popular in the UK and Commonwealth countries. They are often aligned with the ISTQB syllabus but have their own examination style.

### 67.2.4 iSQI Certifications

**iSQI (International Software Quality Institute)** offers several certifications, many of which are recognized globally. Key ones include:

- **Certified Agile Tester (CAT)** – Focuses on testing in agile environments.
- **Certified Professional for Software Testing (CPST)** – A modular program covering various testing topics.
- **Certified Professional for Requirements Engineering (CPRE)** – Useful for test analysts who work closely with requirements.

iSQI exams are typically multiple-choice and can be taken online or at test centers.

### 67.2.5 GAQM (Global Association for Quality Management)

GAQM offers a range of IT certifications, including:

- **Certified Software Tester (CST)** – Similar to ISTQB Foundation.
- **Certified Agile Tester (CAT)** – Agile testing certification.
- **Certified Mobile Tester (CMT)** – Specialized in mobile application testing.

These are less common than ISTQB but still recognized in some regions.

## 67.3 Vendor-Specific Certifications

Vendor certifications validate your ability to use specific testing tools effectively. They are highly valued when your organization or target employers use those tools.

### 67.3.1 Selenium Certifications

Selenium is the most popular open-source test automation framework. However, there is no official certification from the Selenium project. Several training providers offer their own "Certified Selenium Professional" credentials. These typically cover:

- Selenium WebDriver
- Locators (XPath, CSS)
- Frameworks (TestNG, JUnit)
- Page Object Model
- Integration with CI/CD

**Examples:**
- **Guru99 Certified Selenium Professional** – Online exam after completing their course.
- **Edureka Selenium Certification** – Industry-recognized in some circles.
- **Simplilearn Selenium Certification** – Part of their automation testing master's program.

While these certifications lack official backing, they can still demonstrate your knowledge to employers. Many employers prefer hands-on experience over such certifications, but a certification can help you get an interview.

### 67.3.2 Micro Focus (formerly HP) Certifications

Micro Focus offers certifications for its testing tools:

- **UFT (Unified Functional Testing) Certification** – For the popular functional testing tool (formerly QTP).
- **LoadRunner Certification** – For performance testing with LoadRunner.
- **ALM (Application Lifecycle Management) Certification** – For test management with ALM/Quality Center.

These certifications are valuable in enterprises that use Micro Focus tools. They typically require passing a proctored exam and may have prerequisites.

### 67.3.3 Tricentis Tosca Certification

Tricentis Tosca is a leading model-based test automation tool. Tricentis offers a **Tosca Certified Professional** program with multiple levels (Tosca Commander Specialist, Tosca Automation Specialist, etc.). Certification requires completing training and passing a practical exam.

### 67.3.4 IBM Certifications

IBM offers certifications for its testing tools:

- **IBM Rational Functional Tester** certification.
- **IBM Rational Performance Tester** certification.
- **IBM Rational Quality Manager** certification.

These are relevant for organizations using IBM's Rational suite.

### 67.3.5 TestRail Certification

TestRail by Gurock offers a **TestRail Certification** program that validates your ability to use TestRail effectively for test case management. It covers test case organization, test runs, reports, and best practices.

### 67.3.6 Cloud Provider Certifications

Cloud platforms like AWS, Azure, and Google Cloud offer certifications that include testing aspects, particularly in DevOps and CI/CD pipelines.

- **AWS Certified DevOps Engineer – Professional** – Includes testing in CI/CD, infrastructure as code, and monitoring.
- **Microsoft Certified: DevOps Engineer Expert** – Covers testing strategies, Azure Test Plans, and continuous testing.
- **Google Professional Cloud Developer** – Includes testing in cloud applications.

While not purely testing certifications, they demonstrate your ability to integrate testing into cloud-based development.

### 67.3.7 Atlassian Certifications

Atlassian offers certifications for Jira and Confluence, which are relevant for test management (especially when using Jira with Xray or Zephyr). The **Atlassian Certified Jira Administrator** certification can be valuable for test managers who administer test management plugins.

## 67.4 Domain-Specific Certifications

Some certifications focus on specific domains like security, agile, or performance, complementing your testing skills.

### 67.4.1 Security Testing Certifications

- **CSSLP (Certified Secure Software Lifecycle Professional)** – Offered by ISC2, this certification focuses on integrating security throughout the SDLC. It covers security testing, risk management, and secure coding. Requires 4 years of experience in at least one of the CSSLP domains.
- **CISSP (Certified Information Systems Security Professional)** – A broad security certification that includes some testing aspects (e.g., security assessment). More for security professionals than testers.
- **CEH (Certified Ethical Hacker)** – Focuses on penetration testing and ethical hacking, which includes aspects of security testing. Useful for testers specializing in security.

### 67.4.2 Agile Certifications

- **ICAgile Certified Professional in Testing (ICP-TST)** – Offered by ICAgile, this certification focuses on agile testing practices, collaboration, and mindset. It is part of their Agile Testing track.
- **Scrum.org Professional Scrum Developer (PSD)** – Covers development practices in Scrum, including unit testing, integration, and acceptance testing.
- **Scrum Alliance Certified Scrum Developer (CSD)** – Similar to PSD, includes testing topics.

### 67.4.3 Performance Testing Certifications

- **Certified Performance Tester (CPT)** – Offered by various bodies (e.g., QAI). Focuses on performance testing concepts, tools, and analysis.
- **LoadRunner Certified Professional** – As mentioned under Micro Focus.
- **JMeter Certification** – Various providers offer JMeter-specific certifications (e.g., from Udemy, Simplilearn). No official JMeter certification exists, but training providers issue their own.

## 67.6 How to Choose the Right Certification

### 67.6.1 Align with Career Goals

- **Aspiring Test Manager:** Consider CSTE, or ISTQB Advanced Test Manager. Project management certifications like PMP or PRINCE2 can also be beneficial.
- **Test Analyst / Functional Tester:** ISTQB Advanced Test Analyst, BCS Practitioner, or ICAgile ICP-TST are good choices.
- **Automation Specialist:** Focus on tool-specific certifications (Selenium, Tosca, UFT) and maybe ISTQB Advanced Technical Test Analyst.
- **Performance Tester:** LoadRunner, JMeter certifications, or ISTQB Advanced Technical Test Analyst (with performance focus).
- **Security Tester:** CSSLP, CEH, or security certifications like CISSP.

### 67.6.2 Consider Industry Demand

- Check job postings in your region for preferred certifications. In some countries, ISTQB is the standard; in others, CSTE or BCS may be more recognized.
- Vendor certifications are valuable if your target companies use those specific tools.

### 67.6.3 Validate Credibility

- Stick to well-known certifying bodies (QAI, iSQI, BCS, ISC2, AWS, etc.). Avoid "diploma mill" certifications that have little industry recognition.
- Research the certification's reputation on LinkedIn, forums, and with peers.

### 67.6.4 Cost and Time

- Certifications vary widely in cost. Factor in exam fees, training (if needed), and study materials.
- Some certifications require significant experience (e.g., CSSLP, PMP). Ensure you meet prerequisites before investing time.

### 67.6.5 Combining Certifications

Many professionals combine certifications to build a well-rounded profile:

- **Entry Level:** ISTQB Foundation + CAST or BCS Foundation.
- **Generalist:** ISTQB Advanced Test Analyst + CSTE.
- **Automation:** ISTQB Foundation + Certified Selenium Professional + ISTQB Advanced Technical Test Analyst.
- **Management:** ISTQB Advanced Test Manager + PMP + CSTE.
- **Security:** ISTQB Foundation + CSSLP + CEH.

## 67.7 Preparation Resources

- **Official Syllabi and Sample Exams:** Always start with the official materials from the certifying body.
- **Training Courses:** Many providers offer accredited training (online or in-person). These are often expensive but provide structured learning and may include exam vouchers.
- **Books:** Look for study guides specific to the certification. For example, "CSTE Study Guide" by QAI, or "CSSLP All-in-One Exam Guide."
- **Online Practice Tests:** Websites like Udemy, ExamTopics, or the certifying body's own practice exams help you get familiar with the question format.
- **Communities:** Join LinkedIn groups, Reddit communities (e.g., r/softwaretesting), and forums where candidates share experiences and tips.

---

## 67.8 Common Pitfalls

- **Chasing Too Many Certifications:** Certifications are valuable, but experience and practical skills matter more. Don't collect them without applying the knowledge.
- **Choosing Obscure Certifications:** Verify that the certification is recognized in your industry. A certification no one has heard of may not help your career.
- **Ignoring Prerequisites:** Ensure you meet experience or education requirements before registering; otherwise, you may not be eligible or may fail due to lack of knowledge.
- **Neglecting Renewal:** Some certifications require continuing education credits (CEUs) or renewal exams. Factor in maintenance costs and effort.

---

## 67.9 Conclusion

The world of software testing certifications extends far beyond ISTQB. General certifications like CSTE and CAST, vendor-specific credentials for tools like Selenium and LoadRunner, and domain-focused certifications in security, agile, or DevOps all offer pathways to deepen your expertise and advance your career.

When choosing a certification, consider your career goals, industry demand, and the credibility of the certifying body. Remember that certifications complement, but do not replace, practical experience. A balanced approach – gaining hands-on skills, earning relevant certifications, and continuously learning – will position you for success in the dynamic field of software testing.
